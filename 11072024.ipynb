```python
'''
Desafio 1: Crie um programa que realiza a contagem de 1 até 100, usando apenas números ímpares.
Ao final do processo, exiba em tela quantos números ímpares foram encontrados nesse intervalo, assim como a soma dos mesmos.
'''

# Variável que vai iniciar a contagem de números ímpares encontrados no intervalo entre 1 até 100.
cont_impares = 0  # Esta variável vai contar quantos números ímpares existem. Começamos com 0 porque ainda não contamos nenhum número.

# Variável que vai somar os números ímpares encontrados no intervalo entre 1 até 100.
soma_impares = 0  # Esta variável vai acumular a soma dos números ímpares encontrados. Começamos com zero porque ainda não somamos nenhum número.

# Loop for - 1 a 100
for numero in range(1, 101):  # O laço for cria e usa a variável 'numero' automaticamente. Ela começa no 1 e vai até 100 (inclusive).
    # Verifica se o número é ímpar
    if numero % 2 != 0:  # O operador % é o "módulo". Ele retorna o resto da divisão de um número por outro. Se o resto da divisão de um número por 2 não for 0, então esse número é ímpar.
        print(numero)  # Se o número for ímpar, ele é mostrado na tela.
        cont_impares += 1  # Esta linha aumenta o contador de números ímpares em 1, assim que um número ímpar for encontrado. O operador += é uma maneira abreviada de escrever cont_impares = cont_impares + 1.
        soma_impares += numero  # Esta linha adiciona o número ímpar encontrado à soma total. O operador += é uma maneira abreviada de escrever soma_impares = soma_impares + numero.

# Exibe a quantidade de números ímpares encontrados
print(f"O total de números ímpares encontrados no intervalo de 1 a 100 é {cont_impares}.")  # Aqui mostramos quantos números ímpares foram encontrados. A variável cont_impares guarda esse valor.

# Exibe a soma dos números ímpares encontrados
print(f"A soma total dos números ímpares encontrados no intervalo de 1 a 100 é {soma_impares}.")  # Aqui mostramos a soma de todos os números ímpares encontrados. A variável soma_impares guarda esse valor.
```

```python
'''
Desafio 2: Crie uma lista de tarefas básica, onde você pode adicionar quantas tarefas desejar:
    - Use While
    - Use For
    - Use If-Else
    - Use os ('cls')
'''

# Importa a biblioteca para limpar a tela
import os  # Essa biblioteca 'os' é utilizada para executar comandos do sistema operacional. Nesse caso, será usada para limpar a tela.

# Inicializa uma lista vazia para armazenar as tarefas que o usuário adicionar.
tarefas = []

# Loop infinito para permitir a adição de tarefas e continuará rodando até encontrar uma instrução 'break', que no exercicio é identificada pela inserção da resposta 'n'.
while True:
    # Limpa a tela para facilitar a visualização por meio do comando 'cls' que foi executado graças à importação da biblioteca 'os'.
    os.system('cls' if os.name == 'nt' else 'clear')  # os.system é uma função da biblioteca 'os'. Se o sistema for Windows, a limpeza da tela é feita pelo comando 'cls'; se for outro SO, usa-se o comando 'clear'.
    
    # Exibe as tarefas atuais
    print("Tarefas Atuais:")
    for i, tarefa in enumerate(tarefas, start=1):  # 'enumerate' cria um índice automaticamente
        print(f"{i}. {tarefa}")  # Exibe cada tarefa com um número
    
    # Pergunta ao usuário se ele deseja adicionar uma nova tarefa, convertendo a resposta para caixa baixa.
    adicionar = input("Deseja adicionar uma nova tarefa? (s/n): ").lower()
    
    # Se a resposta for 's', adiciona a nova tarefa
    if adicionar == 's':
        nova_tarefa = input("Digite a nova tarefa: ")  # Solicita a nova tarefa ao usuário
        tarefas.append(nova_tarefa)  # Adiciona a nova tarefa à lista 'tarefas'
    else:
        # Se a resposta não for 's', encerra o loop
        break

# Exibe a lista final de tarefas
print("Lista Final de Tarefas:")
for tarefa in tarefas:  # Itera sobre a lista de tarefas
    print(tarefa)  # Exibe cada tarefa
```

```python
'''
Desafio 3: Crie um programa que peça ao usuário digitar um número aleatório, em seguida retorne se esse número é primo ou não, caso não seja, retorne também a quantas vezes esse número é divisível.
Dica: Número primo é aquele que apenas é divisível por 1 ou por ele mesmo, temos de criar essa condição lógica para poder solucionar esse exercício.
'''

# Solicita ao usuário para digitar um número aleatório
numero = int(input("Digite um número: "))  # Converte a entrada do usuário de uma string para um número inteiro

# Inicializa um contador de divisores
contador_divisores = 0  # Esta variável vai contar quantas vezes o número informado pelo usuário é divisível. É iniciada em 0.

# Loop para verificar quantas vezes o número é divisível
for i in range(1, numero + 1):  # O laço for vai de 1 até o número digitado pelo usuário (inclusivo)
    # Se o número for divisível por 'i'
    if numero % i == 0:  # O operador % (módulo) retorna o resto da divisão. Se o resto for 0, então 'numero' é divisível por 'i'
        contador_divisores += 1  # Incrementa o contador de divisores

# Verifica se o número é divisível apenas por 1 e por ele mesmo (primo)
if contador_divisores == 2:  # Um número primo tem exatamente 2 divisores: 1 e ele mesmo
    print(f"{numero} é um número primo.")
else:
    # Se não for primo, exibe a quantidade de divisores
    print(f"{numero} não é um número primo.")
    print(f"Ele é divisível {contador_divisores} vezes.")
```
