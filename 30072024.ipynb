Abaixo seguem as principais categorias de linguagem SQL, cada uma com seu propÃ³sito e comandos especÃ­ficos para gerenciar e manipular dados em um banco de dados relacional.

#### **DDL (Data Definition Language - Linguagem de DefiniÃ§Ã£o de Dados)**
Utilizada para definir e modificar a estrutura dos objetos de banco de dados, como tabelas, Ã­ndices e esquemas.

    * CREATE: Cria um novo objeto no banco de dados, como uma tabela, Ã­ndice ou vista.
    * ALTER: Modifica a estrutura de um objeto existente no banco de dados.
    * DROP: Remove um objeto do banco de dados.
    * TRUNCATE: Remove todos os registros de uma tabela, mas mantÃ©m sua estrutura para uso futuro.
    * COMMENT: Adiciona comentÃ¡rios a um objeto do banco de dados.
    * RENAME: Renomeia um objeto do banco de dados.

```python
-- DDL
CREATE TABLE alunos (
    id INT PRIMARY KEY,
    nome VARCHAR(100),
    idade INT
);

ALTER TABLE alunos ADD COLUMN endereco VARCHAR(255);

DROP TABLE alunos;
```

### **DML (Data Manipulation Language - Linguagem de ManipulaÃ§Ã£o de Dados)**
Utilizada para manipular os dados contidos nos objetos de banco de dados.
    * A DML Ã© usada para manipular os dados contidos nos objetos de banco de dados.
    * Comandos Comuns:
    * INSERT: Insere novos dados em uma tabela.
    * UPDATE: Atualiza dados existentes em uma tabela.
    * DELETE: Remove dados de uma tabela.

```python
-- DML
-- InserÃ§Ã£o de dados em uma tabela
INSERT INTO alunos (id, nome, idade) VALUES (1, 'JoÃ£o', 22);

-- AtualizaÃ§Ã£o de dados em uma tabela
UPDATE alunos SET idade = 23 WHERE id = 1;

-- RemoÃ§Ã£o de dados de uma tabela
DELETE FROM alunos WHERE id = 1;
```

### **DQL (Data Query Language - Linguagem de Consulta de Dados)**
A DQL Ã© usada para consultar os dados contidos nos objetos de banco de dados. O principal comando Ã© o SELECT.
    * SELECT: Recupera dados de uma ou mais tabelas.

```python
-- DQL
SELECT nome, idade FROM alunos WHERE idade > 20;
```

### **DCL (Data Control Language - Linguagem de Controle de Dados)**
A DCL Ã© usada para controlar o acesso aos dados e garantir a seguranÃ§a no banco de dados.
    * GRANT: Concede permissÃµes a um usuÃ¡rio ou papel.
    * REVOKE: Revoga permissÃµes de um usuÃ¡rio ou papel.

```python
-- DCL
-- ConcessÃ£o de permissÃµes a um usuÃ¡rio
GRANT SELECT, INSERT ON alunos TO usuario1;

-- RevogaÃ§Ã£o de permissÃµes de um usuÃ¡rio
REVOKE INSERT ON alunos FROM usuario1;
```
