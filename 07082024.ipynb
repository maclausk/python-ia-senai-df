```python
# Exemplo - Tabela Brasileirão 2024 (Conforme dados de 07/08/2024)
# Copiamos a URL e atribuimos a uma variável, no caso, url. 
url = "https://www.terra.com.br/esportes/futebol/brasileiro-serie-a/tabela/"
tabela_seriea = pd.read_html(url, encoding='UTF-8')
# Cria-se agora uma varíavel tabela_seriea para fazer a leitura e armazenar o conteudo da url
tabela_seriea
```

```python
# Como algumas tabelas estão vazias, iremos dropar as tabelas inválidas. Nesse caso apenas a tabela 'Times.2' é válida. 
tabela_brasileirao = tabela_brasileirao.drop(columns=['Times','Times.1','Times.3'])
```

```python
# Apresentando a tabela após a exclusão das colunas inválidas.
tabela_brasileirao
```

```python
# renomear tabela 'Times.2' para 'Times'

tabela_brasileirao.rename(columns={'Times.2':'Times'}, inplace=True) # inplace altera o dataframe original
tabela_brasileirao
```

```python
# Usando str.replace - Retirar o sinal '>>' e substituir por vazio ou em branco
# REGEX = ele tá negando que o sinal de '>>' é uma expressão regular, mas sim uma string.
# Para remover os espaços em brancos
# Indica que >> é um texto literal, não uma expressão regular.

tabela_brasileirao['Times'] = tabela_brasileirao['Times'].str.replace('>>','',regex=False).str.strip()

```

```python
# Visualizando a tabela corrigida
tabela_brasileirao
```

```python
# O que esse comando faz?
grafico_percent = tabela_brasileirao.groupby('Times')['%'].sum().sort_values()
grafico_percent
```
