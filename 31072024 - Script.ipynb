#### `5. Teoria dos Conjuntos`

A **Teoria dos Conjuntos** da matemÃ¡tica Ã© usada para organizar e relacionar dados em bancos de dados. Ela ajuda a entender como os elementos (ou registros) de diferentes tabelas se conectam entre si.

### Exemplos PrÃ¡ticos

1. **Chaves PrimÃ¡rias (Primary Keys)**:
   - SÃ£o como identificadores Ãºnicos em uma tabela. Por exemplo, o `id` de um aluno na tabela `aluno`.
   - **Exemplo**: Cada aluno tem um `id` Ãºnico. Assim, "JoÃ£o" pode ter `id` 1 e "Maria" pode ter `id` 2.

2. **Chaves Estrangeiras (Foreign Keys)**:
   - SÃ£o usadas para criar uma ligaÃ§Ã£o entre duas tabelas. Elas fazem referÃªncia a uma chave primÃ¡ria em outra tabela.
   - **Exemplo**: A tabela `diariobordo` tem uma coluna `fk_aluno_id` que faz referÃªncia ao `id` na tabela `aluno`. Isso mostra que uma entrada no diÃ¡rio de bordo pertence a um aluno especÃ­fico.

### Relacionamento Entre Tabelas

- **Um para Muitos (1:N)**: Um aluno pode ter muitos registros no diÃ¡rio de bordo.
  - **Exemplo**: Se o `id` do aluno "JoÃ£o" Ã© 1, entÃ£o `fk_aluno_id` no `diariobordo` pode ter vÃ¡rios registros com o valor 1, indicando que todos esses registros pertencem a "JoÃ£o".

### SQL no Uso da Teoria dos Conjuntos

No SQL, usamos comandos como `JOIN`, `FOREIGN KEY`, e `PRIMARY KEY` para implementar essas relaÃ§Ãµes.

```sql
-- CriaÃ§Ã£o da tabela aluno
CREATE TABLE aluno (
    id INT PRIMARY KEY AUTO_INCREMENT,
    ra CHAR(8) NOT NULL
);

-- CriaÃ§Ã£o da tabela diariobordo
CREATE TABLE diariobordo (
    id INT PRIMARY KEY AUTO_INCREMENT,
    texto TEXT NOT NULL,
    datahora DATETIME,
    fk_aluno_id INT,
    FOREIGN KEY (fk_aluno_id) REFERENCES aluno(id)
);
```
