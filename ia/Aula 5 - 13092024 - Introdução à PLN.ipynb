### Introdução ao PLN (Processamento de Linguagem Natural)

O **Processamento de Linguagem Natural (PLN)** é a área da inteligência artificial que ensina os computadores a entenderem, interpretarem e até responderem de forma inteligente à linguagem humana.

---

#### 1. Texto (Corpus)

- **Corpus** é o nome dado ao texto em PLN. Quando temos vários textos, chamamos isso de **corpora**. 
- Exemplo: Se estivermos trabalhando com um único artigo de notícias, ele será o **corpus**. Se estivermos trabalhando com vários artigos, eles serão chamados de **corpora**.

---

#### 2. Tokenização (Separação do Texto)

- A **tokenização** é o processo de dividir um texto em palavras ou frases menores, chamadas de **tokens**.
- **Exemplo**: A frase "Eu adoro programação" pode ser dividida assim:
    - Tokens: ["Eu", "adoro", "programação"]

---

#### 3. Análise Léxica (Stemming)

- A **análise léxica** envolve um processo chamado **stemming**, que é a redução de palavras à sua raiz.
- **Exemplo**: A palavra "entregar" pode ser reduzida à sua raiz "entreg". Isso ajuda o computador a reconhecer que palavras como "entregue" e "entrega" têm o mesmo significado central.

---

#### 4. Análise Sintática

- A **análise sintática** verifica a estrutura gramatical de uma frase, ou seja, como as palavras estão organizadas.
- Isso envolve o uso de **árvores sintáticas** que mostram a relação entre as palavras.
- **Exemplo**: Na frase "O cachorro corre rápido", podemos ver que "O cachorro" é o sujeito, "corre" é o verbo e "rápido" é o advérbio.

---

#### 5. Análise Semântica (Contexto)

- A **análise semântica** foca no significado das palavras dentro de um contexto.
- **Exemplo**: A palavra "banco" pode significar uma instituição financeira ou um assento, dependendo do contexto da frase.
  - "Eu fui ao banco" (instituição financeira).
  - "Eu sentei no banco do parque" (assento).

- ![](https://colband.net.br/files/2013/04/vendo-por-do-sol.jpg)

---

#### 6. Análise Pragmática (Intenção)

- A **análise pragmática** tenta entender a **intenção** por trás das palavras.
- **Exemplo**: Na frase "Você pode abrir a janela?", a pessoa não está apenas perguntando se o outro é capaz de abrir a janela, mas provavelmente pedindo para abrir a janela.

---

#### 7. Significado Pretendido

- Esse conceito trata de entender o que o falante ou autor queria dizer com suas palavras.
- **Exemplo**: Se alguém diz "Está quente aqui dentro", pode estar sugerindo que quer abrir uma janela, mesmo que não tenha dito isso diretamente.

---

### Etapa 2: Tokenização (Palavra ou Sentença)

**Tokenização** é o processo de dividir um texto em unidades menores, chamadas **tokens**. Esses tokens podem ser palavras ou sentenças, dependendo da abordagem escolhida.

Em **Processamento de Linguagem Natural (PLN)**, a tokenização é um passo fundamental para a análise de textos, pois transforma o texto bruto em uma sequência de palavras ou sentenças, permitindo uma análise mais detalhada.

#### Tipos de Tokenização:
1. **Tokenização por Palavra**: Separa o texto em palavras individuais.
   - Exemplo: A frase "Eu adoro programação" seria dividida em: `["Eu", "adoro", "programação"]`.

2. **Tokenização por Sentença**: Separa o texto em sentenças completas.
   - Exemplo: O texto "Eu adoro programação. Python é incrível!" seria dividido em: `["Eu adoro programação.", "Python é incrível!"]`.

---

````python
# Importação das funções para tokenizar palavras e sentenças
import nltk
from nltk.tokenize import word_tokenize  # Tokeniza palavras
from nltk.tokenize import sent_tokenize  # Tokeniza sentenças
````

# O que é um WordCloud?

Uma Word Cloud (ou nuvem de palavras) é uma visualização gráfica que representa a frequência de palavras em um texto. Nesse tipo de visualização, palavras mais frequentes são exibidas em tamanhos maiores, enquanto palavras menos frequentes aparecem em tamanhos menores. Isso permite que você veja rapidamente quais palavras são mais proeminentes em um conjunto de dados textual.

# O que são STOPWORDS?

Em Processamento de Linguagem Natural (PLN), stopwords são palavras comuns que são frequentemente removidas de um texto durante a análise, pois geralmente não carregam muito significado semântico para o objetivo da análise. Essas palavras incluem artigos, preposições, conjunções e outras palavras que são frequentemente usadas em um idioma, mas que têm pouco impacto na compreensão do conteúdo principal do texto.

# Por que as Stopwords são Removidas?

- Redução de Ruído: Stopwords podem adicionar "ruído" aos dados, diluindo a análise de texto com palavras que não são úteis para a compreensão do significado ou para a identificação de padrões.

- Eficiência: Remover stopwords pode ajudar a reduzir o tamanho dos dados e, assim, melhorar a eficiência dos algoritmos de processamento e análise.

- Foco em Palavras Relevantes: A remoção de stopwords permite que você se concentre nas palavras mais significativas que realmente contribuem para o significado do texto

# Criando uma função plot_cloud




````python
# Definição da Função
def plot_cloud(wordcloud):
    # Criar a Figura
    plt.figure(figsize=(40, 30))

    # Exibir a Word Cloud
    plt.imshow(wordcloud)

    # Remover Eixos
    plt.axis("off")
````

## Criando um objeto WordCloud

3.  **random_state**=1: define a semente geradora de números aleatórios (reprodutível na mesma nuvem)
4.  **background_color**='salmon': cor de fundo da imagem da nuvem de palavras
5.  **colormap='Pastel'**: define esquema de cores a ser usado para as palavras na nuvem.
6.  **collocations=False**: define se as palavras compostas (collocations) devem ser exibidas.
7.  **stopwords=STOPWORDS**:conjunto de palavras para serem ignoradas na nuvem de palavras. Deve ser uma coleção de palavras a serem excluídas da nuvem. Pode ser uma lista ou um conjunto de palavras.
8.  **.generate(texto): gera a nuvem de palavras

- Converter para Array: np.array(...) converte a imagem aberta em um array NumPy.
- Atribuir a Variável: O array é armazenado na variável mascara.

## Uso da Conversão

- Esse tipo de conversão é útil em várias situações, como:

- Aplicar Máscaras: Se você estiver trabalhando com nuvens de palavras ou visualizações que requerem máscaras (por exemplo, para definir a forma da nuvem de palavras), você pode carregar uma imagem de máscara e convertê-la em um array NumPy.
- Processamento de Imagens: Facilita o processamento e a manipulação de imagens usando operações matriciais oferecidas pelo NumPy.
