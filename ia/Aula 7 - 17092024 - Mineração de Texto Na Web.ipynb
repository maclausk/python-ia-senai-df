## O que envolve?

### 1. Web Scraping
- **Definição**: Web scraping é o processo automatizado de coleta de informações da web. Usamos scripts ou bibliotecas (como BeautifulSoup, Scrapy ou Selenium) para "raspar" dados de sites, como artigos de notícias, blogs e outros textos de interesse.
- **Objetivo**: Obter o conteúdo textual de uma página da web, que é a base da mineração de texto. O scraping coleta as palavras que estarão disponíveis para análise.

### 2. Mineração de Texto
- **Definição**: Mineração de texto é a análise de dados textuais, que busca padrões, palavras-chave, sentimentos e outros insights a partir de grandes volumes de texto.
- **Objetivo**: Extrair informações úteis dos textos coletados, como tópicos principais, frequência de palavras, sentimentos expressos, etc.
- **Técnicas Comuns**:
  - **Tokenização**: Dividir o texto em palavras ou frases.
  - **Contagem de palavras**: Identificar quais palavras aparecem com maior frequência.
  - **Extração de tópicos**: Determinar os temas ou assuntos predominantes no texto.

### 3. Processamento de Linguagem Natural (PLN)
- **Definição**: O PLN é uma área da Inteligência Artificial que busca entender e interpretar o conteúdo de textos humanos, ajudando as máquinas a processarem a linguagem natural.
- **Objetivo**: Utilizar ferramentas e bibliotecas como NLTK, SpaCy e Transformers para fazer análises mais avançadas, como reconhecimento de entidades (nomes de pessoas, lugares), análise de sentimentos, e categorização de documentos.

## O que você pode aprender na mineração de texto de artigos na internet?

- **Raspagem de dados textuais**: Coletar artigos de notícias ou blogs automaticamente.
- **Análise de conteúdo**: Extrair as informações mais importantes de um artigo (como título, autor, data de publicação).
- **Análise de sentimentos**: Descobrir se o tom do artigo é positivo, negativo ou neutro.
- **Extração de tópicos**: Identificar os principais temas discutidos em diversos artigos.

## Por que isso é útil?

Empresas, jornalistas, pesquisadores e desenvolvedores utilizam mineração de texto em artigos da internet para:
- **Monitorar notícias**: Descobrir rapidamente quais são os tópicos mais discutidos.
- **Análise de tendências**: Identificar tendências emergentes em diferentes indústrias.
- **Pesquisa de mercado**: Analisar opiniões de clientes ou consumidores a partir de comentários e avaliações.
- **Criação de conteúdo automatizado**: Usar os dados coletados para gerar relatórios ou artigos automaticamente.

```python
# Baixamos os recursos necessários do nltk para tokenização e stopwords
nltk.download('punkt')
nltk.download('stopwords')

# Função para extrair o texto de uma página web usando BeautifulSoup
def extrair_texto_da_pagina(url):
    # Fazemos uma requisição HTTP para a URL
    resposta = requests.get(url)

    # Usamos BeautifulSoup para analisar o conteúdo HTML da página
    soup = BeautifulSoup(resposta.content, 'html.parser')

    # Extraímos o texto bruto da página (ajuste conforme necessário)
    texto = soup.get_text()
    return texto

# URL do artigo no site DODF (substitua com a URL específica)
url = 'https://akiradev.netlify.app/posts/python-cientifico-bibliotecas/'  # Substitua com a URL específica do artigo

# Extraímos o texto da página fornecida
texto = extrair_texto_da_pagina(url)

# Tokenizamos o texto em palavras e convertendo-as para minúsculas para uniformidade
tokens = word_tokenize(texto.lower())

# Carregamos as stopwords do NLTK para a língua portuguesa
stop_words = set(stopwords.words('portuguese'))

# Filtramos os tokens para remover stopwords e qualquer coisa que não seja uma palavra
filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]

# Criamos a distribuição de frequências das palavras filtradas
freq_dist = FreqDist(filtered_tokens)

# Exibimos as 10 palavras mais comuns no texto extraído
print(freq_dist.most_common(10))

# Plotamos a distribuição de frequências das 30 palavras mais comuns
plt.figure(figsize=(12, 6))
freq_dist.plot(30, cumulative=False)
plt.show()

# Geramos uma nuvem de palavras com base na distribuição de frequências
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dist)

# Plotamos a nuvem de palavras gerada
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Removemos os eixos para uma visualização mais limpa
plt.show()
```

```python
# Função para extrair o texto bruto de uma página web
def extrair_texto_da_pagina(url):
    # Faz uma requisição HTTP para a URL fornecida e retorna o conteúdo da página.
    resposta = requests.get(url)

    # Analisamos o conteúdo HTML da página usando BeautifulSoup
    # 'html.parser' é o analisador padrão que converte o conteúdo em um objeto BeautifulSoup
    soup = BeautifulSoup(resposta.content, 'html.parser')

    # Retornamos o texto bruto da página, removendo as tags HTML
    return soup.get_text()
```

```python
# URL do artigo no site DODF
url = 'https://akiradev.netlify.app/posts/python-cientifico-bibliotecas/'
```

```python
# Extraímos o texto da página da web utilizando a função extrair_texto_da_pagina
# A variável 'url' deve conter o endereço da página que será processada
texto = extrair_texto_da_pagina(url)

# Tokenizamos o texto extraído, convertendo tudo para minúsculas para uniformidade
# O método 'word_tokenize' divide o texto em palavras (tokens)
tokens = word_tokenize(texto.lower())

# Carregamos as stopwords (palavras irrelevantes) em português utilizando o NLTK
# Stopwords são palavras como "de", "o", "a", que geralmente não são úteis para análise
stop_words = set(stopwords.words('portuguese'))

# Filtramos os tokens removendo as stopwords e palavras que não sejam alfabéticas
# O método 'word.isalpha()' garante que apenas palavras (sem números ou símbolos) sejam mantidas
filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
```

```python
# Calculamos a distribuição de frequência das palavras filtradas utilizando FreqDist.
# FreqDist é uma classe do NLTK que conta quantas vezes cada palavra aparece no texto.
freq_dist = FreqDist(filtered_tokens)

# Exibimos as 10 palavras mais comuns no texto, junto com suas frequências.
print(freq_dist.most_common(10))
```

```python
# Criamos uma nova figura para o gráfico com o tamanho 12x6.
plt.figure(figsize=(12, 6))

# Plotamos a distribuição de frequência das 30 palavras mais comuns no texto.
# 'freq_dist.plot(30)' mostra as 30 palavras mais frequentes e suas respectivas contagens.
# 'cumulative=False' indica que queremos a frequência de cada palavra individualmente, e não de forma cumulativa.
freq_dist.plot(30, cumulative=False)

# Exibimos o gráfico gerado.
plt.show()
```

```python
# Geramos uma nuvem de palavras a partir da distribuição de frequências calculada.
# Definimos o tamanho da imagem da nuvem de palavras como 3000x2000 pixels e a cor de fundo como preta.
# 'colormap' define o esquema de cores da nuvem de palavras, aqui estamos usando 'viridis' para um efeito visual.
wordcloud = WordCloud(width=3000, height=2000, background_color='black', colormap='viridis').generate_from_frequencies(freq_dist)

# Criamos uma nova figura para a nuvem de palavras com o tamanho 10x5.
plt.figure(figsize=(10, 5))

# Exibimos a nuvem de palavras gerada.
# 'bilinear' é uma técnica de interpolação para suavizar a imagem.
plt.imshow(wordcloud, interpolation='bilinear')

# Removemos os eixos da visualização para deixar a nuvem de palavras mais limpa.
plt.axis('off')

# Mostramos o gráfico com a nuvem de palavras.
plt.show()
```
