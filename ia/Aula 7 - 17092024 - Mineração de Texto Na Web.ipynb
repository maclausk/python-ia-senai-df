#### **5. Extração de Conteúdo com Goose3**

```python
# Importação das bibliotecas necessárias
from goose3 import Goose  # Biblioteca para extração de artigos da web
import nltk               # Biblioteca para processamento de linguagem natural
```

```python
# Inicialização do objeto Goose, que será utilizado para extrair o conteúdo da página
g = Goose()
```

```python
# Definição da URL do artigo que será extraído
url = "https://akiradev.netlify.app/posts/python-cientifico-bibliotecas/"
```

```python
# Extração do artigo a partir da URL fornecida
artigo = g.extract(url)
```

#### **6. Processamento do Texto Extraído**

```python
# Download dos recursos necessários do NLTK
nltk.download('punkt')      # Pacote para tokenização de palavras e sentenças
nltk.download('stopwords')  # Lista de stopwords em vários idiomas
```

```python
# Importação da função de tokenização de palavras
from nltk.tokenize import word_tokenize

# Tokenização do texto limpo do artigo, convertendo para minúsculas
tokens = word_tokenize(artigo.cleaned_text.lower(), language='portuguese')
```

```python
# Importação das stopwords em português
from nltk.corpus import stopwords

# Carregamento da lista de stopwords em português
stop_words = set(stopwords.words('portuguese'))

# Filtragem dos tokens, removendo stopwords e caracteres que não são letras
tokens_filtrados = [word for word in tokens if word.isalpha() and word not in stop_words]
```

```python
# Importação da classe FreqDist para cálculo de frequência de palavras
from nltk.probability import FreqDist

# Cálculo da distribuição de frequência das palavras filtradas
freq_dist = FreqDist(tokens_filtrados)

# Impressão das 10 palavras mais comuns
print("Palavras mais comuns:", freq_dist.most_common(10))
```

```python
# Importação da biblioteca para criação de gráficos
import matplotlib.pyplot as plt

# Configuração do tamanho da figura do gráfico
plt.figure(figsize=(12, 6))

# Plotagem do gráfico de distribuição de frequência das 30 palavras mais comuns
freq_dist.plot(30, cumulative=False)

# Exibição do gráfico
plt.show()
```

```python
### 7. Visualização com WordCloud
```

```python
# Gerando a Nuvem de Palavras
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(freq_dist)
```

```python
# Exibindo a Nuvem de Palavras
plt.figure(figsize=(15, 7.5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Importação de bibliotecas necessárias para extração de dados e análise de texto.

# 1 - NLTK (Natural Language Toolkit): Biblioteca poderosa para Processamento de Linguagem Natural (PLN),
# com ferramentas para tokenização, análise de frequência, stopwords, etc.
import nltk

# 2 - 'os': A biblioteca os do Python permite interação com o sistema operacional,
# fornecendo funções para manipulação de arquivos, pastas, e execução de comandos do sistema.
import os

# 3 - pyplot: Importamos 'pyplot' do matplotlib, uma biblioteca poderosa para criação de gráficos em Python,
# que fornece funções para criar gráficos de linha, barras, dispersão e muitos outros tipos de visualizações.
import matplotlib.pyplot as plt

# 4 - requests: Usada para realizar requisições HTTP para acessar páginas da web.
import requests

# 5 - BeautifulSoup: Utilizada para extrair e analisar o conteúdo HTML de páginas web.
from bs4 import BeautifulSoup

# 6 - word_tokenize: Função do NLTK para tokenizar o texto em palavras.
from nltk.tokenize import word_tokenize

# 7 - 'sent_tokenize': Função do NLTK para tokenizar um texto em sentenças.
from nltk.tokenize import sent_tokenize

# 8 - FreqDist: Ferramenta do NLTK para calcular a distribuição de frequências de itens, como palavras em um texto, permitindo analisar quais palavras ou tokens aparecem com mais frequencia.
from nltk.probability import FreqDist

# 9 - stopwords: Conjunto de palavras comuns e irrelevantes, que são removidas do texto para análise mais eficiente.
from nltk.corpus import stopwords

# 10 - WordCloud: Biblioteca para gerar nuvens de palavras a partir de frequências de palavras.
from wordcloud import WordCloud, STOPWORDS

# 11 - Punkt: Contém modelos que permitem dividir um texto em palavras ou sentenças.
nltk.download('punkt')

# 12 - STOPWORDS: Pacote 'stopwords' do NLTK, que contém listas de palavras irrelevantes (stopwords) em diferentes idiomas, como artigos, preposições, conjunções, etc.
# Essas palavras serão filtradas na análise.
nltk.download('stopwords')
```

```python
# Definimos a variável 'url' para armazenar o endereço do site desejado.
url = "https://akiradev.netlify.app/posts/python-cientifico-bibliotecas/"
```

```python
# Definição da Função 'plot_cloud': Cria e exibe uma nuvem de palavras (Word Cloud) utilizando a biblioteca matplotlib.

def plot_cloud(wordcloud): # Essa função exibe uma nuvem de palavras usando a biblioteca matplotlib.
    # Criar a Figura e define o tamanho da figura para a nuvem de palavras.
    plt.figure(figsize=(40, 30))

    # Exibe a imagem da nuvem de palavras
    plt.imshow(wordcloud)

    # Remover Eixos da imagem para que ela apareça de forma mais limpa, sem as bordas e eixos núméricos.
    plt.axis("off")
```

```python
# A função 'word_tokenize' do NLTK tenta tokenizar a string fornecida (no caso, a URL).
# O parâmetro 'language' define o idioma em que a tokenização será feita (português neste caso).
# No entanto, a tokenização de uma URL pode não ser ideal, já que ela contém símbolos especiais.
nltk.word_tokenize(url, language='portuguese')
```

```python
# A função 'sent_tokenize' do NLTK é usada para dividir o texto em sentenças.
# O argumento 'artigo' deve ser o texto extraído do artigo, como 'artigo.cleaned_text', e não o próprio objeto 'artigo'.
# A função retorna uma lista de sentenças extraídas do texto.
sentencas = nltk.sent_tokenize(artigo.cleaned_text)

# Exibe as primeiras 5 sentenças tokenizadas
print(sentencas[:5])
```

```python
# Função para extrair o texto de uma página web usando BeautifulSoup
def extrair_texto_da_pagina(url):
    # Fazemos uma requisição HTTP para a URL
    resposta = requests.get(url)

    # Usamos BeautifulSoup para analisar o conteúdo HTML da página
    soup = BeautifulSoup(resposta.content, 'html.parser')

    # Extraímos o texto bruto da página (ajuste conforme necessário)
    texto = soup.get_text()
    return texto

# URL do artigo
url = 'https://akiradev.netlify.app/posts/python-cientifico-bibliotecas/'

# Extraímos o texto da página fornecida
texto = extrair_texto_da_pagina(url)

# Tokenizamos o texto em palavras e convertendo-as para minúsculas para uniformidade
tokens = word_tokenize(texto.lower())

# Carregamos as stopwords do NLTK para a língua portuguesa
stop_words = set(stopwords.words('portuguese'))

# Filtramos os tokens para remover stopwords e qualquer coisa que não seja uma palavra
filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]

# Criamos a distribuição de frequências das palavras filtradas
freq_dist = FreqDist(filtered_tokens)

# Exibimos as 10 palavras mais comuns no texto extraído
print(freq_dist.most_common(10))

# Plotamos a distribuição de frequências das 30 palavras mais comuns
plt.figure(figsize=(12, 6))
freq_dist.plot(30, cumulative=False)
plt.show()
```

```python
# Criamos uma nova figura com o tamanho 10x5 para exibir a nuvem de palavras
plt.figure(figsize=(10, 5))

# Exibimos a imagem da nuvem de palavras gerada
# 'imshow' exibe a imagem da WordCloud com interpolação 'bilinear' para suavizar a imagem
plt.imshow(wordcloud, interpolation='bilinear')

# Removemos os eixos da visualização para deixar a imagem mais limpa
plt.axis('off')

# Mostramos o gráfico com a nuvem de palavras
plt.show()
```

```python
# Geramos uma nuvem de palavras a partir da distribuição de frequências calculada.
# Definimos o tamanho da imagem da nuvem de palavras como 3000x2000 pixels e a cor de fundo como preta.
# 'colormap' define o esquema de cores da nuvem de palavras, aqui estamos usando 'viridis' para um efeito visual.
wordcloud = WordCloud(width=3000, height=2000, background_color='black', colormap='viridis').generate_from_frequencies(freq_dist)

# Criamos uma nova figura para a nuvem de palavras com o tamanho 10x5.
plt.figure(figsize=(10, 5))

# Exibimos a nuvem de palavras gerada.
# 'bilinear' é uma técnica de interpolação para suavizar a imagem.
plt.imshow(wordcloud, interpolation='bilinear')

# Removemos os eixos da visualização para deixar a nuvem de palavras mais limpa.
plt.axis('off')

# Mostramos o gráfico com a nuvem de palavras.
plt.show()
```

```python
# Importamos 'get_single_color_func' para permitir a personalização das cores na nuvem de palavras
from wordcloud import get_single_color_func

# Exemplo de uma função de cor personalizada que define a cor das palavras na nuvem de palavras
# A função recebe vários parâmetros (word, font_size, etc.) e retorna uma cor RGB específica.
def cor_personalizada(word, font_size, position, orientation, random_state=None, **kwargs):
    # Definimos uma cor personalizada (azul) para todas as palavras usando a notação RGB
    return 'rgb(50, 150, 255)'  # Cor azul em RGB

# Criamos a nuvem de palavras com a função de cor personalizada onde a cor de fundo é branca e usamos a função 'cor_personalizada' para definir a cor das palavras
wordcloud = WordCloud(width=3000, height=2000, background_color='white', color_func=cor_personalizada).generate_from_frequencies(freq_dist)

# Plotamos a nuvem de palavras
plt.figure(figsize=(10, 5))

# Exibimos a nuvem de palavras gerada com interpolação bilinear
plt.imshow(wordcloud, interpolation='bilinear')

# Removemos os eixos da visualização
plt.axis('off')

# Mostramos o gráfico com a nuvem de palavras
plt.show()
```

### **Distribuição de Palavras Com Máscara**

```python
# Comando para instalar a biblioteca NumPy, que é usada para manipulação de arrays e dados numéricos.
!pip install numpy
```

```python
# Importamos a classe Image do módulo PIL (Python Imaging Library) para abrir e manipular imagens.
from PIL import Image

# Importamos a biblioteca NumPy para manipulação de arrays.
import numpy as np

# Carregamos a imagem "blz.png" e a convertemos para um array NumPy.
# Esse array será usado como máscara na criação de uma nuvem de palavras.
mascara = np.array(Image.open("dados/blz.png"))
```

```python
# A variável 'mascara' contém um array NumPy que representa a imagem "python.png".
# Cada valor no array corresponde a um pixel da imagem, e esse array pode ser usado para definir a forma da nuvem de palavras (WordCloud) ao invés de usar uma forma retangular padrão.
mascara
```

```python
# Criamos uma nuvem de palavras utilizando a máscara 'mascara'.
# Definimos o tamanho da nuvem (3000x2000), cor de fundo preta, e o esquema de cores 'rainbow'.
# A máscara (mask=mascara) molda a nuvem de palavras conforme a imagem fornecida.
# 'collocations=False' evita a combinação de palavras frequentes, e 'stopwords' remove palavras irrelevantes.
wordcloud_mascara = WordCloud(width=3000, height=2000, random_state=1, background_color='black', colormap='rainbow', collocations=False, stopwords=STOPWORDS, mask=mascara).generate(texto)

# 'generate(texto)' cria a nuvem de palavras a partir do texto fornecido.
```

```python
# Criamos uma nova figura para exibir a nuvem de palavras.
# O tamanho da figura é definido como 20x10, e a cor de fundo da área ao redor da nuvem de palavras é amarela (facecolor="Yellow").
plt.figure(figsize=(20, 10), facecolor="Yellow")

# Exibimos a nuvem de palavras gerada com a máscara personalizada.
plt.imshow(wordcloud_mascara)

# Removemos os eixos para que apenas a nuvem de palavras seja visível.
plt.axis("off")

# Ajustamos o layout para remover espaçamentos adicionais (pad = 0) ao redor da nuvem de palavras.
plt.tight_layout(pad=0)

# Exibimos a nuvem de palavras na tela.
plt.show()
```
