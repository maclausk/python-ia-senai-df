### `ExplicaÃ§Ã£o Adicional`

#### Relacionamentos

- **1,1 (Um para Um)**:
  - Significa que um registro em uma tabela estÃ¡ relacionado com um e apenas um registro em outra tabela. Por exemplo, se um aluno sÃ³ pudesse ter uma avaliaÃ§Ã£o, o relacionamento seria 1,1.

- **1,n (Um para Muitos)**:
  - Significa que um registro em uma tabela pode estar relacionado com muitos registros em outra tabela. Por exemplo, um aluno pode ter vÃ¡rias avaliaÃ§Ãµes, mas cada avaliaÃ§Ã£o pertence a um Ãºnico aluno. Isso Ã© representado no cÃ³digo pela chave estrangeira `fk_aluno_id` na tabela `avaliacao`.

#### Chaves e Constraints

- **Chave PrimÃ¡ria (PRIMARY KEY)**:
  - Garante que cada registro na tabela tenha um identificador Ãºnico. No cÃ³digo, `id` Ã© a chave primÃ¡ria de cada tabela.

- **Chave Estrangeira (FOREIGN KEY)**:
  - Garante que um valor em uma coluna (ou conjunto de colunas) corresponde a um valor em uma coluna na tabela referenciada. No cÃ³digo, `fk_aluno_id` na tabela `avaliacao` Ã© uma chave estrangeira que referencia `id` na tabela `aluno`.

- **Constraint (RestriÃ§Ã£o)**:
  - Define regras para os dados em uma tabela. A constraint de chave estrangeira adicionada garante a integridade referencial, assegurando que cada `fk_aluno_id` em `avaliacao` corresponde a um `id` em `aluno`.
