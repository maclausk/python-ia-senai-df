#### Exemplo Orientado a Objeto

````python
# Definição de uma classe chamada 'Pessoa'
class Pessoa():
    
    # Método inicializador (construtor) que é chamado quando uma nova instância da classe é criada.
    # O método __init__ é usado para inicializar os atributos do objeto.
    # 'self' é uma referência à instância atual da classe, usada para acessar variáveis que pertencem à classe.
    def __init__(self, nome, idade):
        self.nome = nome  # Atribui o valor do parâmetro 'nome' ao atributo 'nome' da instância.
        self.idade = idade  # Atribui o valor do parâmetro 'idade' ao atributo 'idade' da instância.
    
    # Método 'pega_nome' que retorna o valor do atributo 'nome' da instância.
    def pega_nome(self):
        return self.nome  # Retorna o valor do atributo 'nome'.
    
    # Método 'dobro_idade' que calcula e retorna o dobro da idade da pessoa.
    def dobro_idade(self):
        return self.idade * 2  # Retorna o dobro do valor do atributo 'idade'.

# Criação de uma instância da classe 'Pessoa' com o nome "joao" e a idade não foi passada, o que causaria um erro.
pessoa = Pessoa("joao", 25)  # Aqui, devemos passar dois argumentos: 'nome' e 'idade'.

# Chamada do método 'pega_nome' da instância 'pessoa'.
# Este método retorna o nome da pessoa, que neste caso é "joao".
pessoa.pega_nome()  # Deve retornar 'joao'.
````

### Series x Dataframes

#### Análise Estatística de Dados

### 1. Média

- A média é calculada somando todos os valores de um conjunto de dados e dividindo essa soma pela quantidade de valores presentes.
- Por exemplo, se uma pessoa dormiu em média 6,42 horas por dia na última semana, significa que o total de horas de sono divididas pelos dias foi 6,42.



### 2. Mediana (percentil 50%)

- A mediana é o valor central em um conjunto de dados ordenados em ordem crescente.
- Se houver uma quantidade ímpar de valores, a mediana é o valor no meio; se houver uma quantidade par, a mediana é a média dos dois valores centrais.

**Exemplo:**

Dados ordenados: `0, 6, 7, 7, 7, 8, 10`

- Mediana: `7` (no caso ímpar)
- Mediana: média entre os dois valores centrais (caso par)

### 3. Média vs Mediana

- **Média**: Considera todos os valores, mas pode ser influenciada por outliers (valores fora da curva).
- **Mediana**: Foca no valor central, ignorando outliers, o que pode representar melhor a distribuição central dos dados.


### 4. Variância e Desvio Padrão



- **Variância**: Mede o quão distante os dados estão da média. É calculada subtraindo cada valor da média, elevando o resultado ao quadrado, somando todos esses valores e dividindo pela quantidade de dados.

Variância = (7 - 6,42)² + (8 - 6,42)²  + (6 - 6,42)² + (0 - 6,42)² + ... / 7 = 10,38

- **Desvio Padrão**: A raiz quadrada da variância, que traz a variação dos dados para a mesma ordem de grandeza dos valores originais.

### 5. Percentil

- O percentil divide um conjunto de dados em 100 partes iguais. 
- Por exemplo, o 20º percentil representa os 20% menores valores. 

- **Percentil 90**: Refere-se aos 90% menores valores, e o restante (10%) são os maiores valores.

### 7. Quartil



Os quartis dividem um conjunto de dados em quatro partes iguais:

- **Q1 (25%)**: 25% dos menores valores (percentil 25)
- **Q2 (50%)**: 50% dos menores valores, que corresponde à mediana (percentil 50)
- **Q3 (75%)**: 75% dos menores valores (percentil 75)

**Exemplo de Cálculo:**

Dados: `[13, 17, 25, 28, 33, 41, 65, 69, 72]`

- Q1 = 21 (mediana do primeiro quartil)
- Q2 = 33 (mediana do conjunto total)
- Q3 = 67 (mediana do terceiro quartil)



### 8. Boxplot

- Um boxplot visualiza a distribuição de dados mostrando os quartis e outliers:

- **Q1, Mediana (Q2), Q3**: Representam os valores dos quartis.
- **LS e LI (Limites Superior e Inferior)**: Calculados como `Q3 + 1,5*(Q3-Q1)` e `Q1 - 1,5*(Q3-Q1)` respectivamente. Valores fora desses limites são considerados outliers.

**Exemplo de Boxplot:**

Dados: `(-2, 3, 5, 8, 9, 11, 13, 14, 15, 17, 35)`

- Mediana = 11
- Q1 = 5
- Q3 = 15
- LS = 30, LI = -10

Os outliers são valores fora do intervalo [-10, 30], neste caso, 35.

### 10. Cálculo dos Quartis e Limites Inferiores/Superiores

#### Quartis
- **Q1 (Quartil 1)**: O primeiro quartil (Q1) é o valor que separa os 25% menores dados.
- **Mediana (Q2)**: A mediana é o valor central em um conjunto de dados.
- **Q3 (Quartil 3)**: O terceiro quartil (Q3) representa os 75% menores dados.

#### Limites Superiores e Inferiores
- **LS (Limite Superior)**: Calculado como `Q3 + 1,5*(Q3 - Q1)`. Determina o ponto além do qual os dados são considerados outliers superiores.
- **LI (Limite Inferior)**: Calculado como `Q1 - 1,5*(Q3 - Q1)`. Determina o ponto abaixo do qual os dados são considerados outliers inferiores.

#### Exemplo de Cálculo
Dado o conjunto de números: `-2, 3, 5, 8, 9, 11, 13, 14, 15, 17, 35`.

- **Mediana (Q2)**: 11
- **Q1 (Quartil 1)**: 5
- **Q3 (Quartil 3)**: 15
- **LS (Limite Superior)**: `15 + 1,5*(15 - 5) = 30`
- **LI (Limite Inferior)**: `5 - 1,5*(15 - 5) = -10`

Outliers seriam os valores acima de 30 ou abaixo de -10. No exemplo, 35 seria considerado um outlier.

#### 11. Utilidade do Boxplot
O **Boxplot** é uma ferramenta visual que permite identificar os quartis e os outliers, fornecendo uma visão clara da dispersão dos dados. Ele é especialmente útil em cenários onde é necessário considerar ou eliminar outliers para calcular medidas como média ou desvio padrão.

# Criando o boxplot
- plt.boxplot(dados)
- Criação do Boxplot: plt.boxplot(dados) gera o boxplot para os dados fornecidos.
- Um boxplot mostra:
    - Q1 (Primeiro Quartil): Onde 25% dos dados estão abaixo.
    - Mediana (Q2): O valor central dos dados.
    - Q3 (Terceiro Quartil): Onde 75% dos dados estão abaixo.
    - Limites Inferior e Superior: Identificam outliers, calculados como ( Q1 - 1.5 times text{AIQ} ) e ( Q3 + 1.5 times text{AIQ} ), onde AIQ é a amplitude interquartil ( (Q3 - Q1) ).
    - Outliers: Pontos fora dos limites que são considerados extremos ou atípicos.

### 12. Histograma
- Um histograma é uma representação gráfica que exibe a distribuição de variáveis contínuas.
- As variáveis são representadas no eixo horizontal, enquanto a altura dos retângulos indica a frequência dos valores.



- O salário mínimo é representado no eixo horizontal.
- A altura dos retângulos indica a frequência e é exibida no eixo vertical.



**Caracteristicas do histograma? Quais são?**

- As variáveis são exibidas no eixo horizontal, e a altura dos retângulos reflete a frequência.
- As variáveis são quantitativas contínuas, podendo assumir qualquer valor dentro dos intervalos (classes) definidos.



````python
# Histograma

import matplotlib.pyplot as plt  # Importa a biblioteca matplotlib e a apelida como 'plt', usada para criar gráficos em Python.

dados = [2, 3, 4, 5, 5, 6, 7, 8, 9, 10]  # Define uma lista chamada 'dados' contendo os valores numéricos que serão analisados e plotados no histograma.

plt.hist(dados, bins=5)  # Cria um histograma dos dados, dividindo-os em 5 intervalos ('bins') para mostrar a distribuição de frequências.

plt.title("Histograma de Frequências")  # Adiciona um título ao gráfico, indicando que é um histograma de frequências.

plt.xlabel("Valores")  # Adiciona um rótulo ao eixo X, indicando que os valores representam os dados em si.

plt.ylabel("Frequência")  # Adiciona um rótulo ao eixo Y, indicando que a altura dos retângulos representa a frequência dos valores.

plt.show()  # Exibe o gráfico na tela. Sem este comando, o gráfico não será mostrado.
````

#### Média por Filme (groupby())

````python
# Agrupando o DataFrame df_filmes pela coluna 'filmes_id' e calculando a média das avaliações para cada filme
medias_por_filme = df_filmes.groupby('filme_id').mean()['avaliacao']

# Exibindo as primeiras médias de avaliações por filme
medias_por_filme.head()
````

````python
# Exibir um histograma das médias de avaliações por filme
# O método plot() é usado para criar o gráfico, onde:
    # - kind='hist' especifica que o gráfico será um histograma
    # - color='green' define a cor das barras do histograma como verde
medias_por_filme.plot(kind='hist', color='green')
````

````python
# Importa a biblioteca seaborn para criar gráficos estatísticos
import seaborn as sns

# Cria um boxplot para visualizar a distribuição das médias de avaliações por filme
# O eixo x (horizontal) representa as médias de avaliações dos filmes
sns.boxplot(x=medias_por_filme)

# O boxplot gerado a partir das médias de avaliações dos filmes apresenta a distribuição dessas médias de forma visual, destacando os seguintes pontos:
    # Caixa Azul: Representa a distribuição central das médias de avaliação. A parte inferior da caixa indica o primeiro quartil (Q1, onde 25% das médias são menores), e a parte superior indica o terceiro quartil (Q3, onde 75% das médias são menores). A linha preta dentro da caixa é a mediana (Q2), que representa o ponto central das médias.
    # Bigodes (Linhas Pretas): Mostram a faixa de variação das médias sem considerar os outliers. Eles se estendem dos quartis até os limites calculados como Q1 - 1,5 * (Q3 - Q1) e Q3 + 1,5 * (Q3 - Q1).
    # Outliers (Pontos Circulares): Representam filmes cujas médias de avaliação são consideravelmente diferentes do resto dos dados, estando fora dos limites dos bigodes.
````

````python
# Calcula as estatísticas descritivas da série 'medias_por_filme', que contém as médias de avaliações por filme
medias_por_filme.describe()

# O método describe() fornece uma visão geral estatística dos dados, incluindo:
# - count: O número total de filmes avaliados.
# - mean: A média das médias das avaliações dos filmes.
# - std: O desvio padrão das médias, que indica o quão dispersas as avaliações estão em relação à média.
# - min: A menor média de avaliação entre os filmes.
# - 25%: O primeiro quartil, indicando que 25% das médias de avaliação são menores ou iguais a esse valor.
# - 50%: A mediana ou segundo quartil (Q2), que é o ponto médio das médias de avaliações.
# - 75%: O terceiro quartil, indicando que 75% das médias de avaliação são menores ou iguais a esse valor.
# - max: A maior média de avaliação entre os filmes.
````

````python
# Importa a biblioteca Seaborn, que é usada para criar gráficos estatísticos atraentes e informativos.
import seaborn as sns

# Cria um boxenplot das médias de avaliação por filme.
sns.boxenplot(x=medias_por_filme)

# Boxenplot é útil para representar a distribuição dos dados, destacando tanto os quartis quanto a densidade dos dados nos extremos.
````

````python
# Cria um histograma das médias de avaliação por filme.
# kde=True adiciona uma curva de densidade de kernel sobreposta ao histograma.
# bins=10 divide os dados em 10 intervalos para o histograma.
sns.histplot(medias_por_filme, kde=True, bins=10)
````
