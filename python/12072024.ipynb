````python
# SISTEMA PYTHON - CADASTROS

import os  # Importa a biblioteca 'os' para permitir a execução de comandos do sistema operacional.
import time  # Importa a biblioteca 'time' para permitir o uso de funções relacionadas ao tempo, como delays.

# Listas de exemplo para CPFs e usuários com 10 registros válidos
lista_cpf = ['12345678900', '11122233344', '22233344455', '33344455566', '44455566677']
# Lista que armazena os CPFs cadastrados.
lista_usuarios = ['gomes', 'marcelo', 'avelino', 'maria', 'joao']
# Lista que armazena os nomes dos usuários cadastrados.

os.system('cls')  # Limpa a tela para facilitar a visualização. 'cls' é o comando de limpar tela no Windows. Se fosse em um sistema Unix, usaria 'clear'.
while True:  # Inicia um loop infinito, que continuará até encontrar um 'break'.
    print(30 * '=', 'Bem-vindo ao sistema Python Cadastros', 30 * '=')
    # Imprime uma linha de boas-vindas, com 30 sinais de igual antes e depois do texto.
    print('1. Cadastrar Usuário')
    print('2. Consultar Usuário')
    print('3. Acessar o Sistema')
    print('4. Remover um usuário')
    print('5. Sair')  # Imprime as opções do menu.

    opcao = input('Digite a opção desejada: ')  # Solicita ao usuário que escolha uma das opções do menu.

    # Opção - 1. Cadastrar Usuário
    if opcao == '1':
        os.system('cls')  # Limpa a tela.
        novo_nome = input('Digite o nome que será cadastrado: ')  # Solicita o nome do novo usuário.
        novo_cpf = input('Digite um novo CPF: ')  # Solicita o CPF do novo usuário.

        # Verificar se o CPF já existe na lista.
        if novo_cpf in lista_cpf:
            print('O CPF digitado já existe!')  # Informa que o CPF já está cadastrado.
        else:
            lista_cpf.append(novo_cpf)  # Adiciona o novo CPF à lista de CPFs.
            lista_usuarios.append(novo_nome)  # Adiciona o novo nome à lista de usuários.
            print('Usuário cadastrado com sucesso!')  # Confirma o cadastro.

    # Opção - 2. Consultar Usuário
    elif opcao == '2':
        os.system('cls')  # Limpa a tela.
        print("Lista de Usuários Cadastrados:")  # Informa que será exibida a lista de usuários cadastrados.
        for usuario in lista_usuarios:  # Itera sobre a lista de usuários.
            print(f'Usuário: {usuario}')  # Imprime cada usuário.

    # Opção - 3. Acessar o Sistema
    elif opcao == '3':
        os.system('cls')  # Limpa a tela.
        cpf_login = input('Digite o CPF: ')  # Solicita o CPF para login.
        
        if cpf_login in lista_cpf:  # Verifica se o CPF está na lista de CPFs cadastrados.
            print('Acesso realizado com sucesso!')  # Confirma o acesso.
        else:
            print('Usuário não possui cadastro!')  # Informa que o CPF não está cadastrado.

    # Opção - 4. Remover um usuário
    elif opcao == '4':
        os.system('cls')  # Limpa a tela.
        cpf_remove = input('Digite o CPF a ser excluído: ')  # Solicita o CPF do usuário a ser removido.

        if cpf_remove in lista_cpf:  # Verifica se o CPF está na lista de CPFs cadastrados.
            indice = lista_cpf.index(cpf_remove)  # Obtém o índice do CPF na lista.
            nome = lista_usuarios.pop(indice)  # Remove o nome correspondente da lista de usuários.
            lista_cpf.pop(indice)  # Remove o CPF da lista de CPFs.
            
            print(f'Usuário: {nome} com o CPF {cpf_remove} foi removido com sucesso!')  # Confirma a remoção.

    # Opção - 5. Sair
    elif opcao == '5':
        os.system('cls')  # Limpa a tela.
        print('Saindo do sistema.')  # Informa que o sistema está encerrando.
        time.sleep(3)  # Gera um atraso de 3 segundos antes de encerrar.
        break  # Encerra o loop infinito.

    # Opção inválida
    else:
        os.system('cls')  # Limpa a tela.
        print('Opção inválida. Por favor, escolha uma opção válida.')  # Informa que a opção digitada é inválida.

print('Você saiu do Sistema')  # Mensagem final após sair do loop.
````

````python
# Jogo da Forca

import random  # import random: Importa a biblioteca random que permite gerar valores aleatórios. 
# Neste caso, será usada para selecionar uma palavra aleatória da lista.

# Lista de palavras que podem ser escolhidas para o jogo.
lista_palavras = ['python', 'programacao', 'computador', 'notebook', 'software', 'openai', 'algoritmo', 'programas']

# Escolhe uma palavra aleatória da lista
palavra = random.choice(lista_palavras)  # Escolhe aleatoriamente uma palavra da lista lista_palavras

# Inicializa listas para letras corretas e erradas
letras_corretas = []  # Lista para armazenar as letras corretas adivinhadas pelo usuário
letras_erradas = []  # Lista para armazenar as letras erradas adivinhadas pelo usuário
tentativas = 6  # Número de tentativas que o usuário tem para adivinhar a palavra

# Loop principal do jogo
while True:  # Cria um loop infinito que será interrompido apenas quando o jogo acabar (break).
    # Inicializa a palavra escondida com underlines
    palavra_escondida = ''  # String para armazenar a palavra com letras adivinhadas ou underlines
    for letra in palavra:  # Percorre cada letra da palavra escolhida
        if letra in letras_corretas:  # Se a letra está na lista de letras corretas, adiciona a letra à palavra escondida.
            palavra_escondida += letra  # Adiciona a letra à palavra escondida
        else:
            palavra_escondida += '_'  # Se a letra não está na lista de letras corretas, adiciona um underline à palavra escondida.

    # Exibe o estado atual da palavra, letras erradas e tentativas restantes
    print('Palavra: ', palavra_escondida)  # Mostra a palavra escondida com letras adivinhadas e underlines
    print('Letras erradas:', letras_erradas)  # Mostra as letras erradas adivinhadas até agora
    print('Tentativas restantes:', tentativas)  # Mostra o número de tentativas restantes

    # Verificar se o usuário ganhou ou perdeu
    if palavra_escondida == palavra:  # Se a palavra escondida é igual à palavra original, o jogador ganhou.
        print('Você acertou!')  # Informa que o usuário acertou a palavra
        break  # Sai do loop
    elif tentativas == 0:  # Se as tentativas acabaram, o jogador perdeu.
        print(f'Você perdeu! A palavra era: {palavra}')  # Informa que o usuário perdeu e mostra a palavra correta
        break  # Sai do loop

    # Solicita ao usuário para digitar uma letra
    letra_usuario = input('Digite uma letra: ').lower()  #  Solicita ao usuário uma letra e a converte para minúscula.

    # Verifica se a letra digitada está na palavra
    if letra_usuario in palavra:  # Verifica se a letra está na palavra
        print('Letra correta!')  # Informa que a letra está correta
        letras_corretas.append(letra_usuario)  # Adiciona a letra à lista de letras corretas.
    else:
        print('Letra incorreta!')  # Informa que a letra está incorreta
        letras_erradas.append(letra_usuario)  # Adiciona a letra à lista de letras erradas e decrementa o número de tentativas.
        tentativas -= 1  # Decrementa o número de tentativas restantes
````
