```python
# Importando a biblioteca Pandas para análise de dados e atribuindo a ela o alias 'pd'
import pandas as pd

# Importando os dados do arquivo CSV 'pesquisamusicalv2.csv' para um DataFrame chamado 'df_pm2'
df_pm2 = pd.read_csv('dataset/pesquisamusicalv2.csv')
```

```python
# Exibindo os primeiros registros do DataFrame para visualizar a estrutura dos dados
df_pm2.head()
```

```python
# Exibe os últimos 5 registros do DataFrame 'df_pm2'
df_pm2.tail()
```

```python
# Visualizando Estatísticas Descritivas do Dataframe
df_pm2.describe()
```

```python
# Visualizando as Colunas do DataFrame
df_pm2.columns
```

```python
# Renomeando as Colunas do Dataframe e exibindo-a ao final.
df_pm2.columns = [
    'datahora',
    'rock',
    'hiphop',
    'mpb',
    'reggae',
    'funk',
    'sertanejo',
    'gospel',
    'e-music',
    'samba',
    'forró',
    'trap',
    'professor',
    'faixa_etaria']

df_pm2
```

```python
# Removendo a coluna 'datahora' e atualizando o DataFrame
df_pm2 = df_pm2.drop(columns=['datahora'])

# Exibindo as primeiras linhas do novo DataFrame
df_pm2.head()
```

```python
# Criar um Dataframe para apresentar os estilos musicais
df_estilos = df_pm2[['rock','hiphop','mpb','reggae','funk','sertanejo','gospel','e-music','samba','forró','trap']]
df_estilos
```

```python
# Visualizando Informações Técnicas Sobre o Dataframe
df_pm2.info()
```

```python
# Visualizando as descritivas estatísticas do DataFrame 'df_estilos'
df_estilos.describe()
```

```python
# Criando um DataFrame com a soma das avaliações dos estilos musicais
df_estilos_soma = df_estilos[['rock','hiphop','mpb','reggae','funk','sertanejo','gospel','e-music','samba','forró','trap']].sum()
df_estilos_soma
```

```python
# Calculando a média das avaliações de cada estilo musical
media_estilos = df_estilos.mean()
media_estilos
```

```python
# Calculando a mediana das avaliações, que é uma medida robusta de tendência central, menos influenciada por outliers.
mediana_estilos = df_estilos.median()
mediana_estilos
```

```python
# Importar a biblioteca matplotlib.pyplot para visualização de dados
import matplotlib.pyplot as plt

# Plotar o conteúdo do DataFrame 'df_estilos_soma' em um gráfico de barras
# - 'kind="bar"' especifica que o gráfico será de barras.
# - 'figsize=(10,6)' define o tamanho da figura (10 de largura por 6 de altura).
# - 'color="black"' define a cor das barras como preta.
df_estilos_soma.plot(kind='bar', figsize=(10, 6), color='black')

# Definir o título do gráfico
plt.title('Quantidade de Ocorrências Por Estilo Musical')

# Definir o rótulo do eixo x
plt.xlabel('Estilos Musicais')

# Definir o rótulo do eixo y
plt.ylabel('Quantidade')

# Rotacionar os rótulos no eixo x em 45 graus para melhor legibilidade
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.show()

# Nota: Considerando todos os alunos que realizaram a pesquisa, houve desvantagem para alguns estilos, como forró, trap e samba, pois não estavam presentes no primeiro dia da coleta de dados (21/08).
```

```python
# Obter os nomes únicos dos professores que aparecem na coluna 'professor'
df_pm2['professor'].unique()
```

```python
# Criação da variável 'grupo_professor' que agrupa os estilos musicais votados pelos alunos, organizados por professor. Isso permite ver quais estilos musicais foram mais votados em cada turma de um professor específico.
grupo_professor = df_pm2.groupby('professor')[['rock','hiphop','mpb','reggae','funk','sertanejo','gospel','e-music','samba','forró','trap']]
```

```python
# Visualizando a média das avaliações de estilos musicais por professor
media_por_professor = grupo_professor.mean()
print(media_por_professor)

# Visualizando a soma das avaliações de estilos musicais por professor
soma_por_professor = grupo_professor.sum()
print(soma_por_professor)

# Gerando um gráfico de barras para comparar a soma das avaliações de estilos musicais entre os professores
soma_por_professor.plot(kind='bar', figsize=(10, 6))
plt.title('Soma das Avaliações de Estilos Musicais por Professor')
plt.xlabel('Professor')
plt.ylabel('Soma das Avaliações')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filtrando e armazenando em 'professor_romulo' as avaliações dos estilos musicais feitas pelos alunos da turma do professor Rômulo. Em seguida, exibe as avaliações dessa turma específica.
professor_romulo = grupo_professor.get_group('Rômulo')
professor_romulo
```

```python
# Filtrando apenas as colunas relacionadas aos estilos musicais para o DataFrame 'professor_romulo', que já contém as avaliações dos alunos da turma do professor Rômulo.
professor_romulo = professor_romulo[['rock','hiphop','mpb','reggae','funk','sertanejo','gospel','e-music','samba','forró','trap']]

# Exibindo o DataFrame filtrado
professor_romulo
```

```python
# Exibindo as colunas do DataFrame 'professor_romulo' para visualizar os nomes dos estilos musicais avaliados pelos alunos da turma do professor Rômulo.
professor_romulo.columns
```

```python
# Exibindo estatísticas descritivas das notas atribuídas aos estilos musicais pelos alunos da turma do professor Rômulo.
professor_romulo.describe()
```

```python
# Exibindo informações técnicas sobre o DataFrame da turma do professor Rômulo. Isso inclui o número de entradas, nomes das colunas, tipos de dados e memória ocupada.
professor_romulo.info()
```

```python
# Criando um DataFrame com a soma das avaliações dos estilos musicais da turma do professor Rômulo
professor_romulo_soma = professor_romulo.sum()

# Exibindo a soma das avaliações para cada estilo musical
professor_romulo_soma
```

```python
# Plotando as somas das avaliações dos estilos musicais da turma do professor Rômulo em um gráfico de barras horizontais
# - 'kind="barh"' especifica que o gráfico será de barras horizontais.
# - 'figsize=(10, 6)' define o tamanho da figura (10 de largura por 6 de altura).
professor_romulo_soma.plot(kind='barh', figsize=(10, 6))
```

```python
# Calculando a média das avaliações dos estilos musicais da turma do professor Rômulo
professor_romulo.mean()
```

```python
# Calculando a mediana das avaliações dos estilos musicais da turma do professor Rômulo
professor_romulo.median()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Cores para o gráfico de pizza
colors = sns.color_palette("Set3", len(professor_romulo_soma))

# Configurando o tamanho da figura
plt.figure(figsize=(12, 10))

# Plotando o gráfico de pizza
professor_romulo_soma.plot(kind='pie', autopct='%1.1f%%', colors=colors)

# Configurando o gráfico para ser circular
plt.axis('equal')

# Adicionando a legenda no canto superior direito
plt.legend(loc='upper right')

# Removendo o rótulo do eixo y
plt.ylabel('')

# Adicionando um título ao gráfico
plt.title('Porcentagem Média de Votos por Estilo Musical - Turma Professor Rômulo')

# Exibindo o gráfico
plt.show()
```
