```python
# Criando um dicionário com informações sobre apartamentos.
# Estamos criando um dicionário chamado apartamentos. Em um dicionário, cada item é um par chave-valor.
# Neste caso, as chaves são os nomes das regiões (por exemplo, "taguatinga") e os valores são o número de apartamentos disponíveis em cada região (por exemplo, 3).

apartamentos = {
    "taguatinga": 3,
    "brazlândia": 5,
    "ceilandia": 2
}

# Acessando valores no dicionário
# Utilizamos apartamentos["taguatinga"] para acessar o valor associado à chave "taguatinga". 
# Isso retorna o número de apartamentos disponíveis em Taguatinga, que é 3.
# print(apartamentos["taguatinga"])  # Saída: 3

# Adicionando um novo apartamento.
# Adicionamos uma nova chave ("candangolândia") ao dicionário apartamentos com o valor 4.
# Isso representa a inclusão da informação de que há 4 apartamentos disponíveis em Candangolândia

apartamentos["candangolândia"] = 4

# Iterando sobre as chaves e valores
# Utilizamos um laço for para iterar sobre todos os itens do dicionário.
# O método .items() retorna uma lista de pares chave-valor.
# Para cada iteração, apartamento será a chave (o nome da região) e quantidade será o valor (o número de apartamentos).

for apartamento, quantidade in apartamentos.items():
    print(f'{apartamento}: {quantidade}')
```
