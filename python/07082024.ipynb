* 'Web scraping ou 'raspagem de dados'

Web scraping, ou raspagem de dados, é uma técnica utilizada para extrair informações de websites. Essa prática é especialmente útil quando precisamos obter dados de tabelas, listas ou qualquer conteúdo disponível em páginas web, e integrá-los em nossos projetos para análises mais detalhadas.

**Exemplo - Tabela do Brasileirão**
Vamos utilizar a tabela do Brasileirão como exemplo. Primeiro, precisamos identificar um website que contenha a tabela do campeonato em um formato estruturado, como HTML. Usaremos uma biblioteca Python chamada `pandas` para realizar essa tarefa.

**Por que utilizar Web Scraping com Pandas?**
A biblioteca `pandas` é amplamente utilizada para manipulação e análise de dados. Ela possui funções que permitem ler diretamente tabelas HTML de uma página web e convertê-las em DataFrames, facilitando a análise e manipulação dos dados.

**Passos:**
1. Encontrar um website com a tabela do Brasileirão.
2. Utilizar `pandas` para ler a tabela diretamente da web.
3. Manipular e analisar a tabela utilizando as funções agregadas da biblioteca `pandas`.

Este processo é considerado uma função agregada da biblioteca `pandas` porque combina a leitura de dados da web com a poderosa capacidade de manipulação de dados do `pandas`, proporcionando uma maneira eficiente de extrair e analisar dados web.


```python
# Importar a biblioteca Pandas
import pandas as pd
```

```python
# Bibliotecas necessárias para leitura de páginas HTML para raspagem de dados, deixando o Pandas preparado para a raspagem de dados.
    # lxml é uma biblioteca para processar e analisar arquivos XML e HTML.
    # html5lib é uma biblioteca que cria um parser HTML para transformar documentos HTML e XML em árvores de análise.
    # beautifulsoup4 é uma biblioteca para extrair dados de arquivos HTML e XML. Ela cria uma árvore de análise para facilitar a navegação e a busca de elementos.

! pip install lxml
! pip install html5lib
! pip install beautifulsoup4
```

```python
# Exemplo - Tabela Brasileirão 2024 (Conforme dados de 07/08/2024)

# Copiamos a URL do website que possui a tabela do Campeonato Brasileiro de Futebol 2024, apelidado de Brasileirão
# e atribuímos a uma variável chamada 'url'.
url = "https://www.terra.com.br/esportes/futebol/brasileiro-serie-a/tabela/"

# Utilizamos a função 'read_html' do Pandas para ler a tabela da URL e armazenamos o conteúdo na variável 'tabela_seriea'.
# O parâmetro 'encoding' é utilizado para garantir que a leitura do conteúdo seja feita com a codificação correta.
tabela_seriea = pd.read_html(url, encoding='UTF-8')

# Exibimos a tabela armazenada na variável 'tabela_seriea' para verificar o conteúdo capturado.
tabela_seriea
```

```python
# Exibe o conteúdo da variável 'tabela_seriea' para visualização
# 'tabela_seriea' utilizando a função display do Jupyter Notebook, que é útil para visualizar de forma clara o conteúdo da variável, especialmente quando se trata de dados tabulares ou listas de DataFrames.
display(tabela_seriea)

# Verifica o tipo de dado da variável 'tabela_seriea'
type(tabela_seriea)  # Resultado esperado: list
```

```python
tabela_brasileirao = tabela_seriea[0]
# Esta linha de código seleciona o primeiro DataFrame da lista tabela_seriea.
# Quando utilizamos pd.read_html, ele retorna uma lista de DataFrames, uma vez que uma página HTML pode conter múltiplas tabelas.
# Aqui, estamos assumindo que a primeira tabela da lista é a tabela que nos interessa, por isso selecionamos o índice [0]
```

```python
# Exibe o conteúdo do DataFrame 'tabela_brasileirao'
# Quando executada em um ambiente Jupyter Notebook, o DataFrame será mostrado em uma forma tabular, permitindo a visualização dos dados contidos nele.
tabela_brasileirao
```

```python
# Mostra o tipo de dados gerados anteriormente. Esta linha de código é utilizada para mostrar o tipo de dados do objeto tabela_brasileirao.
type(tabela_brasileirao)

# O comando type() retorna o tipo do objeto passado como argumento.
# No contexto deste código, tabela_brasileirao foi extraída de uma lista de DataFrames, portanto, o resultado esperado é <class 'pandas.core.frame.DataFrame'>.
```

```python
# Esta linha de código é utilizada para exibir as últimas 4 linhas do DataFrame tabela_brasileirao.
tabela_brasileirao.tail(4)

# O método .tail() é uma função do Pandas que retorna as últimas n linhas (no caso, 4) do DataFrame.
# No contexto do Campeonato Brasileiro de Futebol (Brasileirão), as últimas 4 posições da tabela frequentemente correspondem aos times que estão na zona de rebaixamento, ou seja, aqueles que correm o risco de serem rebaixados para uma divisão inferior na próxima temporada.
```

```python
# Esta linha de código é utilizada para exibir as primeiras 10 linhas do DataFrame tabela_brasileirao.
# O método .head() é uma função do Pandas que retorna as primeiras n linhas (no caso, 10) do DataFrame.
tabela_brasileirao.head(10)

# No contexto do Campeonato Brasileiro de Futebol (Brasileirão), as primeiras 10 posições da tabela frequentemente correspondem aos times que estão no topo da competição.
# incluindo os que estão disputando o título e aqueles que estão em posições de classificação para competições internacionais.
```

```python
# Verificando a quantidade de tabelas e linhas da tabela. Esta linha de código retorna a dimensão do DataFrame
# O método .shape é uma propriedade do Pandas que retorna uma tupla representando a dimensão do DataFrame.
tabela_brasileirao.shape  # Resultado esperado: (20, 13)
```

```python
# Esta linha de código retorna o índice das colunas do DataFrame tabela_brasileirao.
# O método .columns é uma propriedade do Pandas que retorna uma Index contendo os nomes das colunas do DataFrame.
tabela_brasileirao.columns
```

```python
# Como algumas tabelas estão vazias, ou possuem informações irrelevantes, iremos dropar as tabelas inválidas.
# Nesse caso, remove as colunas Times, Times.1, e Times.3 do DataFrame tabela_brasileirao, sendo a 'Times.2' é válida.
# Ao remover colunas irrelevantes, garantimos que nosso DataFrame contém apenas dados úteis e evitamos qualquer confusão ou erro nas etapas subsequentes de análise.
tabela_brasileirao = tabela_brasileirao.drop(columns=['Times', 'Times.1', 'Times.3'])
```

```python
# Apresentando a tabela após a exclusão das colunas inválidas.
tabela_brasileirao
```

```python
# gera um resumo estatístico descritivo do DataFrame.
# A estatística descritiva é uma ferramenta essencial na análise de dados, proporcionando uma visão geral das propriedades dos dados.
# É útil para entender a distribuição, tendência central e dispersão dos dados
tabela_brasileirao.describe()
```

```python
# Retorna informações gerais do Dataframe como o número de entradas (linhas), o nome e tipo de cada  coluna, e a quantidade de valores não nulos em cada coluna.
tabela_brasileirao.info()
```

```python
# Esta linha exibe o índice das colunas do DataFrame.
# Saber os nomes das colunas é essencial para entender a estrutura do DataFrame e para realizar operações específicas em colunas desejadas.
tabela_brasileirao.columns
```

```python
# Mostra o tipo de dados gerados anteriormente. Nesse caso o resultado será um DataFrame.
# Confirmar o tipo do objeto após a realização de operações é uma boa prática para garantir que os dados foram manipulados corretamente e estão no formato esperado para futuras operações de análise e manipulação.
type(tabela_brasileirao)
```

```python
# Renomear coluna 'Times.2' para 'Times'
# 'inplace=True' faz com que a alteração seja feita diretamente no DataFrame original, sem a necessidade de atribuí-lo a uma nova variável.
tabela_brasileirao.rename(columns={'Times.2':'Times'}, inplace=True) # inplace altera o DataFrame original
tabela_brasileirao
```

```python
# Substitui '>>' por uma string vazia, indicando que '>>' é um texto literal, uma string, e negando que seja uma expressão regular (regex=False).
# .str.strip(): Remove espaços em branco no início e no final das strings na coluna 'Times'.
tabela_brasileirao['Times'] = tabela_brasileirao['Times'].str.replace('>>','',regex=False).str.strip()
```

```python
# Este comando exibe o DataFrame tabela_brasileirao após as correções terem sido aplicadas, permitindo verificar se as alterações foram realizadas corretamente.
tabela_brasileirao
```

```python
# groupby('Times'): Agrupa os dados pelo nome dos times.
# ['%'].sum(): Calcula a soma da coluna '%' para cada grupo (time).
# sort_values(): Ordena os valores resultantes em ordem crescente.
# grafico_percent: Armazena o resultado da operação, que é uma série contendo a soma dos valores da coluna '%' para cada time, ordenada em ordem crescente.


grafico_percent = tabela_brasileirao.groupby('Times')['%'].sum().sort_values()
grafico_percent
```

```python
# Este comando instala a biblioteca matplotlib, que é amplamente utilizada para criar gráficos e visualizações em Python.
! pip install matplotlib
```

```python
# Importa a biblioteca matplotlib para criação de gráficos
import matplotlib.pyplot as plt
```

```python
# Cria um gráfico de barras horizontais
# kind='barh' especifica que o gráfico será de barras horizontais
# figsize define o tamanho da figura
# color define a cor das barras
grafico_percent.plot(kind='barh', figsize=(14, 10), color='black')
plt.title('Aproveitamento dos Times da Serie A')  # Adiciona um título ao gráfico
plt.xlabel('Porcentagem')  # Adiciona um rótulo ao eixo x
plt.ylabel('Times')  # Adiciona um rótulo ao eixo y
plt.show()  # Exibe o gráfico
```
