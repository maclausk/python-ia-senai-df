##### Passo 4: Criar Ambiente Virtual

- Um ambiente virtual isola as dependências do seu projeto, evitando conflitos entre diferentes versões de pacotes. 
- No VSCode, use a barra de pesquisa para criar um ambiente virtual ou use o comando "Control+Shift+P", onde deve ser escolhida a opção '.venv' e em seguida a linguagem Python. 

Para ativar o ambiente virtual: Execute o comando 'activate' no terminal, via prompt de comando, seguindo o seguinte exemplo: '\.venv\Scripts\activate'.



Para desativar o ambiente virtual: Execute o comando 'activate' no terminal, via prompt de comando, seguindo o seguinte exemplo: '\.venv\Scripts\activate'.

##### 7: Observações Importantes

**Notebooks:** São como cadernos digitais onde você pode escrever e executar código passo a passo. Eles ajudam a combinar explicações e gráficos em um só lugar, tornando o aprendizado e a análise de dados mais fáceis e interativos.

**Por que usar um ambiente virtual?**
- **Isolamento**: Garante que os pacotes e suas versões sejam específicos para cada projeto, evitando problemas com outros projetos.
- **Organização**: Facilita a gestão de dependências, mantendo seu ambiente de desenvolvimento limpo e organizado.

**Por que desativar o ambiente virtual?**
- **Recursos**: Libera memória e processamento quando não está usando o projeto.
- **Flexibilidade**: Permite alternar para outros projetos que podem ter suas próprias configurações.
- **Consistência**: Assegura que o projeto use as mesmas versões de pacotes sempre que você trabalhar nele.
- **Evitar Conflitos**: Previne problemas com pacotes de outras partes do sistema.
