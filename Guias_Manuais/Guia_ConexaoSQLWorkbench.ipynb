### Configurao da Conexo no MySQL Workbench

---

1. **Abra o MySQL Workbench**.
2. Na tela principal, clique em **+** ao lado de "MySQL Connections" para adicionar uma nova conexo.
3. Preencha os campos conforme indicado:
   - **Connection Name**: `senai_ia_vespertino` (ou outro nome de sua preferncia)
   - **Hostname**: `127.0.0.1`
   - **Port**: `3306`
   - **Username**: `root`
   - **Password**: Clique em "Store in Vault" e insira a senha `senai@123`.
4. Deixe o campo **Default Schema** em branco, ou preencha com o nome do banco de dados que ser utilizado, se j estiver disponvel.
5. Aps preencher os campos, clique em **Test Connection** para verificar se a conexo foi estabelecida com sucesso.
6. Se o teste de conexo for bem-sucedido, clique em **OK** para salvar a nova conexo.
7. Agora, voc pode utilizar essa conexo para executar scripts SQL.

**Nota**: Verifique se o servidor MySQL est rodando localmente ou no IP configurado, e se as credenciais so vlidas.

### Criando o Diagrama EER no MySQL Workbench

1. **Abra o MySQL Workbench** e conecte-se ao banco de dados usando a conexo configurada anteriormente (`senai_ia_vespertino`).
2. Siga os seguintes passos para gerar o diagrama:

   - V at o menu **Database** e selecione **Reverse Engineer**.
   - Escolha a conexo que deseja usar e clique em **Next**.
   - Na prxima janela, escolha o banco de dados que voc deseja modelar (no caso, `semanatextmining`) e clique em **Next**.
   - O Workbench ir importar as tabelas do banco de dados para o modelo EER.
   - Aps a importao, clique em **Next** e depois em **Finish** para visualizar o Diagrama EER (Modelo de Relacionamento Entidade-Relacionamento).

3. **Gerando o Diagrama**:
   - O diagrama EER ser gerado automaticamente com todas as tabelas e suas relaes.
   - Para ajustar o layout do diagrama, arraste e reposicione as tabelas manualmente at que o diagrama fique claro e bem organizado.

4. **Exportando o Diagrama**:
   - Aps o diagrama estar como desejado, v at **File** > **Export** > **Export as PNG** para salvar uma imagem do diagrama.
   
5. Utilize a imagem gerada para documentao ou incluso em relatrios.



**Nota**: Certifique-se de que todas as tabelas e suas chaves estrangeiras foram corretamente importadas para o diagrama EER.
