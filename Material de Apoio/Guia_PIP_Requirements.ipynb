### `PIP - Pip Installs Packages | Requirements.txt`

##### `1. PIP - PIP INSTALLS PACKAGES`

- O pip  um acrnimo para "Pip Installs Packages". Ele  o gerenciador de pacotes oficial para o Python. Com o pip, voc pode instalar, atualizar e remover pacotes Python que so distribudos via o Python Package Index (PyPI) e outras fontes.



- O pip  uma ferramenta essencial para qualquer desenvolvedor Python, pois facilita o processo de configurao e gerenciamento de bibliotecas e dependncias em projetos Python.

`1.1. Principais funes do pip:`
- Instalao de pacotes: Voc pode usar pip para instalar pacotes Python diretamente do PyPI, um repositrio centralizado de bibliotecas Python.
- Gerenciamento de dependncias: Quando voc instala um pacote com pip, ele tambm instala automaticamente todas as dependncias necessrias para que o pacote funcione corretamente.
- Atualizao de pacotes: Voc pode atualizar os pacotes instalados para suas verses mais recentes usando pip.
- Desinstalao de pacotes: pip tambm permite que voc remova pacotes instalados que no so mais necessrios.

`1.2. Exemplos de uso:`


- Instalar um pacote:

    ```
    pip install nome_do_pacote
    ```

* Atualizar um pacote:

    ```
    pip install --upgrade nome_do_pacote
    ```

* Desinstalar um pacote:

    ```
    pip uninstall nome_do_pacote
    ```

* Listar pacotes instalados:

    ```
    pip list
    ```

* Atualizar o pip para a verso mais recente

    ```
    python -m pip install --upgrade pip
    ```

##### `2. REQUIREMENTS.TXT:`



- O arquivo *requirements.txt*  um arquivo de texto simples que lista as dependncias (bibliotecas e pacotes) que um projeto Python precisa para funcionar corretamente.
- Ele desempenha um papel crucial na gesto de dependncias e na criao de ambientes de desenvolvimento e produo reprodutveis.
- Detalhes do conteudo do arquivo requirements.txt:
    - O arquivo requirements.txt contm uma lista de pacotes Python, cada um em uma linha, seguido opcionalmente pela verso especfica necessria. Cada linha segue um formato padro.

    - Exemplo: `nome-do-pacote==verso`
        - numpy==1.21.0
        - pandas==1.3.0
        - matplotlib==3.4.2

    - Voc tambm pode listar apenas o nome do pacote, sem especificar a verso:
        - flask
        - requests

`2.1. Propsito e Benefcios:`
- Reprodutibilidade: O principal benefcio do requirements.txt  garantir que o ambiente de desenvolvimento seja reprodutvel. Isso significa que qualquer pessoa que use o projeto pode instalar exatamente as mesmas verses dos pacotes necessrios.
- Facilidade de Instalao: Com o requirements.txt, um novo desenvolvedor ou um ambiente de produo pode instalar todas as dependncias necessrias simplesmente executando:

    ```
    pip install -r requirements.txt
    ```
- Controle de Verso: Especificar as verses dos pacotes evita que mudanas em verses futuras quebrem o cdigo existente, pois garante que as verses usadas so aquelas testadas e confirmadas como funcionando corretamente.


`2.2. Criando um requirements.txt:`

Se voc j tem um ambiente virtual (venv) configurado com todos os pacotes instalados, ara criar o *requirements.txt* com as dependncias atuais do seu projeto, voc geralmente executa o comando a seguir.

```
   pip freeze > requirements.txt
```

- A criao de um arquivo requirements.txt dentro de uma virtual environment (venv)  opcional, mas altamente recomendada.
- Ele serve para listar todas as dependncias do seu projeto, facilitando a replicao do ambiente em outras mquinas e garantindo a instalao correta dos pacotes.
- Embora seja comum gerar o requirements.txt dentro da venv,  recomendvel mant-lo no diretrio raiz do projeto, fora da venv, para facilitar o compartilhamento e controle de verso.

`2.3. Uso Tpico`

- Instalao de Pacotes: Quando algum clona o repositrio do seu projeto, eles podem configurar o ambiente de desenvolvimento com todas as dependncias necessrias usando:

```
pip install -r requirements.txt
```
- Atualizao de Dependncias: Se novos pacotes forem adicionados ou atualizados, voc pode atualizar o requirements.txt e compartilhar essa atualizao com a equipe

`2.4. Boas Prticas`

- Manter o Arquivo Atualizado: Sempre que voc adicionar, remover ou atualizar um pacote, lembre-se de atualizar o requirements.txt.
- Uso de Verses Especficas:  uma boa prtica especificar as verses dos pacotes para evitar problemas de compatibilidade.

`2.5. Exemplo Real`

- Suponha que voc est desenvolvendo uma aplicao web usando Flask e precisa de algumas bibliotecas adicionais para manipulao de dados e autenticao. Seu requirements.txt pode se parecer com isso:

    - Flask==2.0.1
    - SQLAlchemy==1.4.22
    - requests==2.26.0
    - numpy==1.2

- Qualquer desenvolvedor que queira trabalhar nesse projeto pode simplesmente clonar o repositrio e rodar:

```
pip install -r requirements.txt
````

- Isso garantir que todas as dependncias necessrias sejam instaladas na verso correta.
