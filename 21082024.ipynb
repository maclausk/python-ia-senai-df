### Importando a Biblioteca Pandas Para Análise e Manipulação de Dados

```python
import pandas as pd

# Lendo o arquivo CSV '\dataset\estilos_musicais.csv' e armazenando os dados em um DataFrame chamado 'df_estilos_musicais'
pesquisa_musical = pd.read_csv('dataset/pesquisa_musical.csv')

# Exibindo as primeiras linhas do DataFrame 'pesquisa_musical' para verificar os dados carregados
pesquisa_musical.head()
```

```python
# Verificação do Tipo de Objeto - Exibe o tipo do objeto armazenado na variável 'pesquisa_musical' confirmando que o DataFrame foi carregado corretamente.
type(pesquisa_musical)
```

```python
# Remover a coluna 'Carimbo de data/hora' do Dataframe pois essa informação não é mais necessária para as análises subsequentes.
df_nota = pesquisa_musical.drop(columns=['Carimbo de data/hora'])

# Exibir as primeiras linhas do DataFrame atualizado para verificar a remoção da coluna
df_nota.head()
```

```python
# Renomear as colunas do DataFrame para nomes mais simples e intuitivos
df_nota = df_nota.rename(columns={
    'Avalie os estilos musicais [rock]': 'Rock',
    'Avalie os estilos musicais [hip hop]': 'HipHop',
    'Avalie os estilos musicais [mpb]': 'MPB',
    'Avalie os estilos musicais [reggae]': 'Reggae',
    'Avalie os estilos musicais [funk]': 'Funk',
    'Avalie os estilos musicais [sertanejo]': 'Sertanejo',
    'Avalie os estilos musicais [gospel]': 'Gospel',
    'Avalie os estilos musicais [eletrônica]': 'Eletrônica'
})

# Exibir as primeiras linhas do DataFrame atualizado para verificar as renomeações
df_nota.head()
```

```python
# Calcular a soma das pontuações para cada estilo musical
soma_pontuacoes = df_nota.sum()

# Exibir a soma das pontuações para verificar quais estilos receberam mais pontos
soma_pontuacoes
```

```python
# Verificação do Tipo de Objeto Armazenado na Variável 'soma_pontuacoes'.
type(soma_pontuacoes)  # Esperado: <class 'pandas.core.series.Series'>
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico de barras horizontal usando os dados da 'soma_pontuacoes'.
soma_pontuacoes.plot(kind='barh')

# Adicionar título e rótulos aos eixos
plt.title('Soma das Pontuações por Estilo Musical')  # Título do gráfico
plt.xlabel('Pontuação Total')  # Rótulo do eixo x
plt.ylabel('Estilo Musical')  # Rótulo do eixo y

# Exibir o gráfico
plt.show()
```

### Preparo dos Dados Para Gráficos Boxplot e Histograma

```python
# Converter a Series 'soma_pontuacoes' para um DataFrame
# O método .reset_index() transforma a Series em um DataFrame, onde o índice original se torna uma coluna.
df_estilos = soma_pontuacoes.reset_index()

# Renomear as colunas do DataFrame para nomes mais descritivos
df_estilos.columns = ['estilo', 'pontuacao']

# Exibir o DataFrame resultante para verificar a conversão e renomeação das colunas
print(df_estilos)
```

```python
# Contar o número de estilos musicais presentes no DataFrame
df_estilos.estilo.count()
```

```python
# Acessar a Coluna 'estilo' do DataFrame 'df_estilos'.
df_estilos.estilo
```

```python
# Calcular a soma total das pontuações para todos os estilos musicais do DataFrame 'df_estilos'.
df_estilos['pontuacao'].sum()
```

```python
# Agrupar os dados no DataFrame 'df_estilos' pela coluna 'estilo' e calcular a soma das pontuações, ordenando em ordem decrescente.
data_grafico = df_estilos.groupby('estilo').sum()['pontuacao'].sort_values(ascending=False)

# Exibir o resultado para verificar as pontuações totais de cada estilo em ordem decrescente
data_grafico
```

```python
# Agrupamento por 'Estilo' e soma das pontuações, ordenando de forma crescente
dados_grafico = df_estilos.groupby('estilo').sum()['pontuacao'].sort_values()

# Exibir o resultado
dados_grafico
```

```python
# Exibir estatísticas descritivas das pontuações agrupadas por estilo
dados_grafico.describe()
```

```python
# Criar um gráfico do tiop histograma usando os dados de 'dados_grafico' para visualizar a distribuição das pontuações dos estilos musicais.
dados_grafico.plot(kind='hist')

# Adicionar título e rótulos aos eixos
plt.title('Distribuição das Pontuações por Estilo Musical')  # Título do histograma
plt.xlabel('Pontuação Total')  # Rótulo do eixo x
plt.ylabel('Frequência')  # Rótulo do eixo y

# Exibir o histograma
plt.show()
```

```python
# Contar a frequência de cada nota (1 a 5) em todas as colunas do DataFrame 'df_nota'
# O comando abaixo aplica 'value_counts' a cada coluna para contar as notas, preenche valores ausentes com 0,
# e soma as contagens ao longo das colunas, resultando na frequência total de cada nota em todo o DataFrame.
notas_contagem = df_nota.apply(pd.Series.value_counts).fillna(0).sum(axis=1).astype(int)

# Exibir a contagem de notas (ordenada) para verificar a distribuição das notas
print("Contagem de notas (1 a 5):")
print(notas_contagem.sort_values())

# Armazenar as contagens ordenadas em 'dados_grafico' para possíveis visualizações
dados_grafico = notas_contagem.sort_values()

# Exibir 'dados_grafico' para confirmar os valores ordenados
dados_grafico
```

```python
# Criar um histograma para a distribuição das contagens de notas (1 a 5) usando os dados de 'notas_contagem'.
notas_contagem.plot(kind='hist')

# Adicionar título e rótulos aos eixos
plt.title('Distribuição das Notas Atribuídas (1 a 5)')  # Título do histograma
plt.xlabel('Nota')  # Rótulo do eixo x
plt.ylabel('Frequência')  # Rótulo do eixo y

# Exibir o histograma
plt.show()
```

### Histograma para Exibir a Distribuição das Notas por Estilo Musical

```python
import matplotlib.pyplot as plt

# Definir o tamanho da figura para o histograma
plt.figure(figsize=(14, 10))

# Criar subplots para cada estilo musical
estilos = ['Rock', 'HipHop', 'MPB', 'Reggae', 'Funk', 'Sertanejo', 'Gospel', 'Eletrônica']
for i, estilo in enumerate(estilos, 1):
    plt.subplot(4, 2, i)  # Configura os subplots em uma grade de 4x2
    plt.hist(df_nota[estilo], bins=10, color='skyblue', edgecolor='black')  # Cria o histograma para o estilo
    plt.title(estilo, fontsize=12)  # Adiciona o título com o nome do estilo musical
    plt.xlabel('Notas')  # Rótulo do eixo x
    plt.ylabel('Frequência')  # Rótulo do eixo y

# Ajusta o layout para evitar sobreposição
plt.tight_layout()

# Exibe o histograma
plt.show()
```

#### Explicação do Gráfico Boxplot

- Um gráfico `boxplot` no contexto apresentado é uma ferramenta estatística que mostra a distribuição das notas atribuídas aos diferentes estilos musicais. Ele resume a distribuição dos dados destacando os valores mínimos, máximos, a mediana, os quartis e possíveis outliers.

**Componentes do Boxplot:**

1. **Mediana (Linha Central da Caixa)**:
   - A linha central dentro da caixa representa a mediana das notas, ou seja, o valor central quando as notas são ordenadas. Metade das notas está acima desse valor e a outra metade está abaixo.
   - A mediana é um indicador do valor típico atribuído a um estilo musical.

2. **Quartis (Caixa)**:
   - A caixa no gráfico representa o intervalo interquartil (IQR), que é o intervalo entre o primeiro quartil (Q1) (25% das menores notas) e o terceiro quartil (Q3) (25% das maiores notas). Essa caixa mostra onde estão concentradas as notas do meio 50%.
   - A largura da caixa (IQR) indica a variabilidade das notas para cada estilo. Caixas mais largas indicam maior dispersão das notas, enquanto caixas mais estreitas sugerem que as notas estão mais concentradas em um intervalo específico.

3. **Extremos (Bigodes)**:
   - As linhas que se estendem a partir da caixa, conhecidas como bigodes, indicam o menor e o maior valor das notas que não são considerados outliers. Eles se estendem até o valor mínimo (valor mais baixo dentro de 1,5 vezes o IQR abaixo de Q1) e o valor máximo (valor mais alto dentro de 1,5 vezes o IQR acima de Q3).

4. **Outliers (Pontos Fora da Caixa)**:
   - Os outliers são valores que estão fora do intervalo esperado (muito acima ou muito abaixo da maioria dos dados). No gráfico, esses valores são representados por pontos isolados fora dos bigodes. Eles indicam notas que são significativamente diferentes das outras e podem ser considerados irrelevantes ou excepcionais.
   - Os pontos fora dos bigodes representam notas que estão muito distantes do restante dos dados e podem refletir avaliações excepcionais, tanto positivas quanto negativas.

5. **Valores Mínimos e Máximos**:
   - Os extremos dos bigodes representam as notas mais baixas e mais altas, excluindo os outliers.

**Exemplos do Gráfico:**
- **Rock e Eletrônica**: Ambos têm uma ampla distribuição de notas (bigodes longos) e apresentam a mediana próxima ao topo, indicando que a maioria das notas são altas, mas também possuem algumas notas baixas.
- **MPB**: Apresenta um outlier claro com uma nota significativamente baixa, representado por um ponto isolado abaixo do bigode inferior. Isso indica que uma ou poucas avaliações deram notas muito mais baixas do que as demais.
- **Funk**: Tem uma caixa relativamente curta e bigodes curtos, sugerindo uma menor variabilidade nas notas atribuídas a este estilo musical.

### Gráfico de Barras para Exibir as Notas Médias de Cada Estilo Musical

```python
import matplotlib.pyplot as plt

# Calcular a média das notas para cada estilo musical
media_notas_estilos = df_nota[['Rock', 'HipHop', 'MPB', 'Reggae', 'Funk', 'Sertanejo', 'Gospel', 'Eletrônica']].mean()

# Criar um gráfico de barras com as médias das notas
plt.figure(figsize=(10, 6))  # Define o tamanho da figura do gráfico
ax = media_notas_estilos.plot(kind='bar', color='skyblue', edgecolor='black')

# Adicionar um título ao gráfico
plt.title('Média das Notas por Estilo Musical', fontsize=16)

# Adicionar o rótulo do eixo y
plt.ylabel('Média das Notas', fontsize=14)

# Adicionar o rótulo do eixo x
plt.xlabel('Estilos Musicais', fontsize=14)

# Rotacionar os rótulos do eixo x para melhor visualização
plt.xticks(rotation=45, ha='right')

# Destacar o estilo musical mais popular
max_value = media_notas_estilos.max()
max_style = media_notas_estilos.idxmax()
ax.text(x=media_notas_estilos.index.get_loc(max_style), y=max_value + 0.1, 
        s=f'{max_style} (Mais Popular)', fontsize=12, color='green', ha='center')

# Destacar o estilo musical menos popular
min_value = media_notas_estilos.min()
min_style = media_notas_estilos.idxmin()
ax.text(x=media_notas_estilos.index.get_loc(min_style), y=min_value - 0.2, 
        s=f'{min_style} (Menos Popular)', fontsize=12, color='red', ha='center')

# Exibir o gráfico
plt.show()
```

### Análise Comparativa: Boxplot vs. `pd.describe()`



### **Contextualização**

O slide apresenta uma comparação entre o uso de **Boxplot** e a função `pd.describe()` da biblioteca Pandas no contexto de **Estatística Descritiva**. A estatística descritiva é crucial para resumir e compreender as características principais de um conjunto de dados.

### **Contexto da Pesquisa**

- **População**: Refere-se ao total de indivíduos em um grupo específico que se deseja estudar. Neste exemplo, a população é de **12 milhões de pessoas**, que é o número total de habitantes da cidade de São Paulo.
  
- **Amostra**: A amostra é uma porção da população que é selecionada para análise. No slide, a amostra utilizada é composta por **1.092 pessoas**. Esta amostra foi usada para uma pesquisa de intenção de voto para a Prefeitura de São Paulo, como indicado pelo recorte de notícia apresentado.

### **Comparação Entre Boxplot e `pd.describe()`**

- **Boxplot**:
    - É uma ferramenta gráfica que permite visualizar a distribuição dos dados. Ele mostra a mediana, os quartis, e os possíveis outliers de forma intuitiva. O boxplot é especialmente útil para entender a dispersão dos dados e identificar valores atípicos que podem influenciar a análise.

- **`pd.describe()`**:
    - A função `describe()` do Pandas fornece uma visão estatística geral dos dados, incluindo a média, desvio padrão, mínimo, máximo, e os percentis. É uma maneira rápida de obter uma compreensão dos aspectos estatísticos essenciais do conjunto de dados.

### **Aplicação Prática**

No contexto da pesquisa de intenção de voto:

- **Boxplot** poderia ser usado para visualizar a distribuição das intenções de voto, permitindo identificar se há candidatos com suporte excepcionalmente alto ou baixo (outliers).
  
- **`pd.describe()`** forneceria um resumo estatístico das intenções de voto, incluindo a média de intenção de voto para cada candidato, a dispersão dessas intenções, e outras estatísticas descritivas chave.

### **Conclusão**

Ambas as ferramentas, Boxplot e `pd.describe()`, são complementares e oferecem insights valiosos sobre os dados. O Boxplot é mais visual e útil para detectar padrões e outliers, enquanto `pd.describe()` é mais estatístico e fornece uma visão resumida dos dados em números.

Essa análise combinada ajuda a formar uma base sólida para a tomada de decisões informadas, especialmente em um contexto onde a compreensão clara dos dados pode influenciar escolhas estratégicas, como em uma pesquisa eleitoral.

### Análise de Variáveis e a Função `groupby()`



### **Contextualização do Slide**

- Este slide apresenta uma categorização das variáveis utilizadas em análises de dados e destaca a função `groupby()` da biblioteca Pandas, que é uma ferramenta poderosa para agrupar e sumarizar dados com base nessas variáveis.
- Essa análise permite que pesquisadores e analistas de dados identifiquem padrões, tendências, e façam comparações significativas entre diferentes grupos dentro de um conjunto de dados.

### **Tipos de Variáveis**

- **Variáveis Quantitativas**:
    - Representam números e podem ser subdivididas em dois tipos:
        - **Contínuas**: Variáveis que podem assumir qualquer valor dentro de um intervalo. Exemplo: Altura, Salário, Inflação.
        - **Discretas**: Variáveis que assumem valores inteiros. Exemplo: Idade, Número de Empregados, Produção de Veículos.

- **Variáveis Qualitativas**:
    - Representam categorias e podem ser divididas em:
        - **Nominais**: Categorias sem ordem intrínseca. Exemplo: Raça/Cor, Sexo, Profissão.
        - **Ordinais**: Categorias com uma ordem intrínseca. Exemplo: Escolaridade, Faixa Etária, Ranking de Reclamações.

### **Aplicação da Função `groupby()`**

- A função `groupby()` no Pandas é utilizada para agrupar dados com base em uma ou mais variáveis, o que permite realizar operações agregadas, como a média, soma, contagem, entre outras, sobre cada grupo formado.
  
- **Exemplo Prático**:
    - Se tivermos um DataFrame com dados sobre salários (uma variável quantitativa contínua) e queremos calcular a média salarial por profissão (uma variável qualitativa nominal), podemos usar a função `groupby()` para agrupar os dados por profissão e, em seguida, aplicar a função de agregação desejada, como `mean()`.

A função `groupby()` combinada com funções agregadas, como `mean()`, é uma ferramenta poderosa para sumarizar e entender grandes conjuntos de dados. Ela permite que os analistas transformem dados detalhados em insights acionáveis, facilitando a tomada de decisões informadas.

### Exploração de Dados: Estatísticas Descritivas e Visualização de Boxplot



### **Contextualização do Slide**

Este slide ilustra duas formas diferentes de explorar e visualizar os resultados de uma análise de dados. Ele destaca a combinação de estatísticas descritivas, obtidas com o método `describe()`, e a visualização dos dados através de um gráfico de Boxplot. Ambas as abordagens são essenciais para entender a distribuição e as características dos dados.

### **Forma 1 – DataFrame ou Series com `describe()`**

- **Estatísticas Descritivas**:
    - O comando `medias_por_filme.describe()` gera uma tabela que resume as principais estatísticas descritivas da Series `medias_por_filme`.
    - **Métricas**:
        - **count**: O número total de filmes avaliados (9.724).
        - **mean**: A média das avaliações (3,26).
        - **std**: O desvio padrão das avaliações, que indica a variabilidade das notas (0,87).
        - **min**: A nota mínima atribuída (0,5).
        - **25%** (primeiro quartil): 25% das avaliações estão abaixo de 2,8.
        - **50%** (mediana): A mediana das avaliações é 3,42.
        - **75%** (terceiro quartil): 75% das avaliações estão abaixo de 3,91.
        - **max**: A nota máxima atribuída (5,0).

- **Utilidade**:
    - As estatísticas descritivas fornecem uma visão geral quantitativa dos dados, ajudando a identificar tendências, dispersão e a centralidade das avaliações.
    - Esta forma é essencial para entender os valores numéricos precisos e as propriedades estatísticas dos dados, permitindo análises mais detalhadas.

### **Forma 2 – Data Visualization com Boxplot**

- **Visualização Gráfica**:
    - O gráfico de Boxplot oferece uma representação visual da distribuição das avaliações.
    - **Componentes do Boxplot**:
        - **Mediana**: A linha no meio da caixa representa a mediana (valor central) das avaliações.
        - **Quartis**: As extremidades da caixa representam o primeiro (Q1) e o terceiro quartil (Q3), mostrando onde está a maioria dos dados.
        - **Bigodes**: As linhas que se estendem da caixa indicam o intervalo dos dados, excluindo outliers.
        - **Outliers**: Os pontos fora dos bigodes indicam valores atípicos que se destacam da distribuição geral.

- **Utilidade**:
    - O Boxplot é extremamente útil para identificar a dispersão dos dados e detectar outliers. Ele visualiza de forma clara a mediana e a variabilidade dos dados, além de fornecer uma compreensão imediata das características da distribuição.

### **Conclusão**

A combinação de estatísticas descritivas e visualização gráfica proporciona uma análise completa dos dados. As estatísticas numéricas fornecem detalhes precisos, enquanto o Boxplot permite uma compreensão visual da distribuição e dos outliers. Juntas, essas abordagens ajudam a formar uma base sólida para a análise e interpretação dos dados, permitindo insights mais profundos e decisões informadas.
