#### git add e git add .

- O comando git add . é utilizado no Git para adicionar todos os arquivos novos ou modificados no diretório atual e seus subdiretórios ao índice (staging area). Isso significa que todos os arquivos que foram alterados, adicionados ou deletados desde o último commit serão preparados para serem incluídos no próximo commit.

Aqui está o que cada parte do comando significa:

'git add': Este é o comando básico para adicionar arquivos ao índice do Git.
. '.' (ponto): Este símbolo representa o diretório atual. Quando usado com git add, ele instrui o Git a adicionar todos os arquivos no diretório atual e em todos os seus subdiretórios.

- Em resumo, git add . prepara todos os arquivos no diretório atual e seus subdiretórios para serem confirmados (committed) no próximo commit.

### Aula - 04/07/2024 - Estrutura de Decisão (if / else)

O conceito de if / else na linguagem de programação Python é usado para fazer decisões no código. Ele permite que você execute blocos de código diferentes com base em certas condições.

#### Estrutura Básica
A estrutura básica de um if / else em Python é a seguinte:

```sh
if condição:
    # Bloco de código a ser executado se a condição for verdadeira
else:
    # Bloco de código a ser executado se a condição for falsa
```

Explicação
if condição: - Avalia uma expressão (condição). Se a condição for verdadeira (True), o bloco de código indentado imediatamente após o if será executado.
else: - Este bloco de código é executado se a condição do if for falsa (False)

Exemplo Simples
Vamos ver um exemplo simples para entender melhor:

```sh
idade = 18

if idade >= 18:
    print("Você é maior de idade.")
else:
    print("Você é menor de idade.")
```

### Explicação do Exemplo

- idade = 18 define uma variável idade com o valor 18.
- if idade >= 18: verifica se o valor da variável idade é maior ou igual a 18.
    - Se a condição for verdadeira, print("Você é maior de idade.") será executado.
    - Se a condição for falsa, print("Você é menor de idade.") será executado.
- Uso do elif
- Além do if e else, você também pode usar elif (abreviação de "else if") para adicionar mais condições:

```sh
nota = 85

if nota >= 90:
    print("A nota é A.")
elif nota >= 80:
    print("A nota é B.")
elif nota >= 70:
    print("A nota é C.")
else:
    print("A nota é D.")
```

### Explicação do Exemplo com elif
- nota = 85 define uma variável nota com o valor 85.
- if nota >= 90: verifica se a nota é maior ou igual a 90. Se verdadeiro, imprime "A nota é A."
- elif nota >= 80: verifica se a nota é maior ou igual a 80. Se verdadeiro, imprime "A nota é B."
- elif nota >= 70: verifica se a nota é maior ou igual a 70. Se verdadeiro, imprime "A nota é C."
- else: se nenhuma das condições anteriores for verdadeira, imprime "A nota é D."

### Resumo
- if: Executa o bloco de código se a condição for verdadeira.
- elif: Adiciona mais condições se a condição if anterior for falsa.
- else: Executa o bloco de código se todas as condições anteriores forem falsas.
- O uso de if / else é essencial para controlar o fluxo do programa e tomar decisões com base em diferentes condições.

---

````python
# Estrutura Condicional - Forma Implícita (mais comum e recomendada)

programador = False

if programador:
    print('if é verdadeiro')
else:
    print("if falso")

print('finalizou o codigo')

# if programador: verifica se programador é True.
# Em Python, valores booleanos (True e False) podem ser usados diretamente em condições.
# if programador: é mais conciso e idiomático em Python.

# Vantagens da Forma Implícita
# - Simplicidade: Menos código para escrever e ler.
# - Conformidade com PEP 8: A PEP 8, o guia de estilo oficial do Python, recomenda o uso da forma implícita.
#- Consistência: Em Python, muitas condições utilizam a forma implícita, o que torna o código mais idiomático.
````

````python
# Estrutura Condicional - Forma Explícita (menos comum)

programador = False

if programador == True:
    print('if é verdadeiro')
else:
    print("if falso")

print('finalizou o codigo')

# if programador == True: verifica explicitamente se programador é igual a True.
# Embora funcione, é considerado redundante porque a comparação explícita não é necessária.
````

### Desafio 3 - Boletim Escolar Complexo

- Crie um sistema para coletar os dados do aluno (nome, idade e cpf).
- Em seguida solicite as notas dos 4 bimestres da disciplina de Python que o aluno teve para verificar a MEDIA.
    - Media para aprovação >= 7
    - Media para recuperaçao >= 5
    - Media para reprovação =< 5

- Para retirar o boletim de notas, o aluno precisa ser maior de idade (18) e caso não seja, somente os pais podem retirar o boletim.
    - Imprima as informações do aluno.
    - Imprima as notas do aluno em cada bimestre.
    - imprima a situação do aluno (Aprovado / Reprovado / Recuperação)
    - Imprima se o boletim será retirado pelos pais ou pelo próprio aluno.

````python
# Desafio 3 - Boletim Escolar Complexo

# Cabeçalho do Boletim
# print(60 * '-', 'BOLETIM ESCOLAR', 60 * '-'): Imprime uma linha com 60 traços, a frase "BOLETIM ESCOLAR", seguida de mais 60 traços.
print(60 * '-', 'BOLETIM ESCOLAR', 60 * '-')

# Inserção de dados do aluno
# Solicita o nome do aluno e converte para maiúsculas. Função: '.upper()'
nome = input('Digite seu nome: ').upper()

# Solicita a idade do aluno
idade = input('Digite sua idade: ')

# Solicita o CPF do aluno (somente números)
cpf = input('Digite seu número de CPF (somente números): ')

# Conversão da idade para inteiro (valores recebidos por input são strings por padrão)
idade = int(idade)

# Inserção das notas da matéria de Python referente aos 4 bimestres
nt1 = input('Digite a nota da matéria de Python referente ao 1º bimestre: ')
nt2 = input('Digite a nota da matéria de Python referente ao 2º bimestre: ')
nt3 = input('Digite a nota da matéria de Python referente ao 3º bimestre: ')
nt4 = input('Digite a nota da matéria de Python referente ao 4º bimestre: ')

# Conversão das notas de string para float
# Quando você recebe entradas do usuário usando o comando input() em Python, as entradas são armazenadas como strings (textos).
# Para realizar operações matemáticas, como calcular a média das notas, você precisa converter essas strings em números.
# A conversão para float é apropriada porque as notas podem ser valores decimais (por exemplo, 8.5, 9.3).
nt1 = float(nt1)
nt2 = float(nt2)
nt3 = float(nt3)
nt4 = float(nt4)

# Cálculo da média das notas
media = (nt1 + nt2 + nt3 + nt4) / 4

# Impressão do boletim escolar
# Imprime o nome, idade, CPF, notas e a média do aluno.
print(30 * '-', "BOLETIM ESCOLAR", 30 * '-')
print(f'Aluno: {nome}')
print(f'Idade: {idade}')
print(f'CPF: {cpf}')
print(f'Nota 1: {nt1}')
print(f'Nota 2: {nt2}')
print(f'Nota 3: {nt3}')
print(f'Nota 4: {nt4}')
print(f'Média das notas: {media:.2f}')  # :.2f para mostrar a média com 2 casas decimais

# Verificação do status do aluno com base na média
# Usa if / elif / else para verificar se o aluno foi aprovado, está de recuperação ou reprovado.
if media >= 7:
    # Se a média for maior ou igual a 7, o aluno está aprovado
    print('Parabéns, você foi aprovado!')
elif media >= 5:
    # Se a média for maior ou igual a 5, mas menor que 7, o aluno está de recuperação
    print('Eita, você está de recuperação!')
else:
    # Se a média for menor que 5, o aluno está reprovado
    print('Que pena! você está reprovado.')

# Se o aluno for menor de idade, exibir mensagem específica
if idade < 18:
    print("As notas só serão entregues a um responsável!")

# Rodapé do Boletim
# Imprime uma linha com 60 traços, a frase "BOAS FÉRIAS", seguida de mais 60 traços.
print(60 * '-', 'BOAS FÉRIAS', 60 * '-')
````
