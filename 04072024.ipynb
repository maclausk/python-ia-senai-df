Aqui est o que cada parte do comando significa:

- Em resumo, git add . prepara todos os arquivos no diretrio atual e seus subdiretrios para serem confirmados (committed) no prximo commit.

### Aula - 04/07/2024 - Estrutura de Deciso (if / else)

O conceito de if / else na linguagem de programao Python  usado para fazer decises no cdigo. Ele permite que voc execute blocos de cdigo diferentes com base em certas condies.

#### Estrutura Bsica
A estrutura bsica de um if / else em Python  a seguinte:

```sh
if condio:
    # Bloco de cdigo a ser executado se a condio for verdadeira
else:
    # Bloco de cdigo a ser executado se a condio for falsa
```

Explicao
if condio: - Avalia uma expresso (condio). Se a condio for verdadeira (True), o bloco de cdigo indentado imediatamente aps o if ser executado.
else: - Este bloco de cdigo  executado se a condio do if for falsa (False)

Exemplo Simples
Vamos ver um exemplo simples para entender melhor:

```sh
idade = 18

if idade >= 18:
    print("Voc  maior de idade.")
else:
    print("Voc  menor de idade.")
```

### Explicao do Exemplo

- idade = 18 define uma varivel idade com o valor 18.
- if idade >= 18: verifica se o valor da varivel idade  maior ou igual a 18.
    - Se a condio for verdadeira, print("Voc  maior de idade.") ser executado.
    - Se a condio for falsa, print("Voc  menor de idade.") ser executado.
- Uso do elif
- Alm do if e else, voc tambm pode usar elif (abreviao de "else if") para adicionar mais condies:

```sh
nota = 85

if nota >= 90:
    print("A nota  A.")
elif nota >= 80:
    print("A nota  B.")
elif nota >= 70:
    print("A nota  C.")
else:
    print("A nota  D.")
```

### Explicao do Exemplo com elif
- nota = 85 define uma varivel nota com o valor 85.
- if nota >= 90: verifica se a nota  maior ou igual a 90. Se verdadeiro, imprime "A nota  A."
- elif nota >= 80: verifica se a nota  maior ou igual a 80. Se verdadeiro, imprime "A nota  B."
- elif nota >= 70: verifica se a nota  maior ou igual a 70. Se verdadeiro, imprime "A nota  C."
- else: se nenhuma das condies anteriores for verdadeira, imprime "A nota  D."

### Resumo
- if: Executa o bloco de cdigo se a condio for verdadeira.
- elif: Adiciona mais condies se a condio if anterior for falsa.
- else: Executa o bloco de cdigo se todas as condies anteriores forem falsas.
- O uso de if / else  essencial para controlar o fluxo do programa e tomar decises com base em diferentes condies.

---

````python
# Estrutura Condicional - Forma Implcita (mais comum e recomendada)

programador = False

if programador:
    print('if  verdadeiro')
else:
    print("if falso")

print('finalizou o codigo')

# if programador: verifica se programador  True.
# Em Python, valores booleanos (True e False) podem ser usados diretamente em condies.
# if programador:  mais conciso e idiomtico em Python.

# Vantagens da Forma Implcita
# - Simplicidade: Menos cdigo para escrever e ler.
# - Conformidade com PEP 8: A PEP 8, o guia de estilo oficial do Python, recomenda o uso da forma implcita.
#- Consistncia: Em Python, muitas condies utilizam a forma implcita, o que torna o cdigo mais idiomtico.
````

````python
# Estrutura Condicional - Forma Explcita (menos comum)

programador = False

if programador == True:
    print('if  verdadeiro')
else:
    print("if falso")

print('finalizou o codigo')

# if programador == True: verifica explicitamente se programador  igual a True.
# Embora funcione,  considerado redundante porque a comparao explcita no  necessria.
````

### Desafio 3 - Boletim Escolar Complexo

- Crie um sistema para coletar os dados do aluno (nome, idade e cpf).
- Em seguida solicite as notas dos 4 bimestres da disciplina de Python que o aluno teve para verificar a MEDIA.
    - Media para aprovao >= 7
    - Media para recuperaao >= 5
    - Media para reprovao =< 5

- Para retirar o boletim de notas, o aluno precisa ser maior de idade (18) e caso no seja, somente os pais podem retirar o boletim.
    - Imprima as informaes do aluno.
    - Imprima as notas do aluno em cada bimestre.
    - imprima a situao do aluno (Aprovado / Reprovado / Recuperao)
    - Imprima se o boletim ser retirado pelos pais ou pelo prprio aluno.

````python
# Desafio 3 - Boletim Escolar Complexo

# Cabealho do Boletim
# print(60 * '-', 'BOLETIM ESCOLAR', 60 * '-'): Imprime uma linha com 60 traos, a frase "BOLETIM ESCOLAR", seguida de mais 60 traos.
print(60 * '-', 'BOLETIM ESCOLAR', 60 * '-')

# Insero de dados do aluno
# Solicita o nome do aluno e converte para maisculas. Funo: '.upper()'
nome = input('Digite seu nome: ').upper()

# Solicita a idade do aluno
idade = input('Digite sua idade: ')

# Solicita o CPF do aluno (somente nmeros)
cpf = input('Digite seu nmero de CPF (somente nmeros): ')

# Converso da idade para inteiro (valores recebidos por input so strings por padro)
idade = int(idade)

# Insero das notas da matria de Python referente aos 4 bimestres
nt1 = input('Digite a nota da matria de Python referente ao 1 bimestre: ')
nt2 = input('Digite a nota da matria de Python referente ao 2 bimestre: ')
nt3 = input('Digite a nota da matria de Python referente ao 3 bimestre: ')
nt4 = input('Digite a nota da matria de Python referente ao 4 bimestre: ')

# Converso das notas de string para float
# Quando voc recebe entradas do usurio usando o comando input() em Python, as entradas so armazenadas como strings (textos).
# Para realizar operaes matemticas, como calcular a mdia das notas, voc precisa converter essas strings em nmeros.
# A converso para float  apropriada porque as notas podem ser valores decimais (por exemplo, 8.5, 9.3).
nt1 = float(nt1)
nt2 = float(nt2)
nt3 = float(nt3)
nt4 = float(nt4)

# Clculo da mdia das notas
media = (nt1 + nt2 + nt3 + nt4) / 4

# Impresso do boletim escolar
# Imprime o nome, idade, CPF, notas e a mdia do aluno.
print(30 * '-', "BOLETIM ESCOLAR", 30 * '-')
print(f'Aluno: {nome}')
print(f'Idade: {idade}')
print(f'CPF: {cpf}')
print(f'Nota 1: {nt1}')
print(f'Nota 2: {nt2}')
print(f'Nota 3: {nt3}')
print(f'Nota 4: {nt4}')
print(f'Mdia das notas: {media:.2f}')  # :.2f para mostrar a mdia com 2 casas decimais

# Verificao do status do aluno com base na mdia
# Usa if / elif / else para verificar se o aluno foi aprovado, est de recuperao ou reprovado.
if media >= 7:
    # Se a mdia for maior ou igual a 7, o aluno est aprovado
    print('Parabns, voc foi aprovado!')
elif media >= 5:
    # Se a mdia for maior ou igual a 5, mas menor que 7, o aluno est de recuperao
    print('Eita, voc est de recuperao!')
else:
    # Se a mdia for menor que 5, o aluno est reprovado
    print('Que pena! voc est reprovado.')

# Se o aluno for menor de idade, exibir mensagem especfica
if idade < 18:
    print("As notas s sero entregues a um responsvel!")

# Rodap do Boletim
# Imprime uma linha com 60 traos, a frase "BOAS FRIAS", seguida de mais 60 traos.
print(60 * '-', 'BOAS FRIAS', 60 * '-')
````
