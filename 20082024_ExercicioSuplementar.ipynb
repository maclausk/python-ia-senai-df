### Gráfico de Pizza Para Quem Bebe

```python
# Contagem de quem bebe e quem não bebe
bebe = df_pesquisa_medica['Bebe'].value_counts()

# Gráfico de Pizza
plt.figure(figsize=(6, 6))
plt.pie(bebe, labels=bebe.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Quem Bebe e Não Bebe')
plt.show()
```

### Gráfico de Pizza Para Quem Bebe e Fuma

```python
# Criando uma nova coluna para categorizar quem bebe e fuma
df_pesquisa_medica['Bebe e Fuma'] = df_pesquisa_medica.apply(
    lambda row: 'Bebe e Fuma' if row['Bebe'] != 'Não' and row['Fuma'] != 'Não' 
    else 'Só Bebe' if row['Bebe'] != 'Não' and row['Fuma'] == 'Não'
    else 'Só Fuma' if row['Bebe'] == 'Não' and row['Fuma'] != 'Não'
    else 'Não Bebe e Não Fuma', axis=1)

# Contagem das categorias
bebe_fuma = df_pesquisa_medica['Bebe e Fuma'].value_counts()

# Gráfico de Pizza
plt.figure(figsize=(6, 6))
plt.pie(bebe_fuma, labels=bebe_fuma.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição de Quem Bebe e Fuma')
plt.show()
```
