### Grfico de Pizza Para Quem Bebe

```python
# Contagem de quem bebe e quem no bebe
bebe = df_pesquisa_medica['Bebe'].value_counts()

# Grfico de Pizza
plt.figure(figsize=(6, 6))
plt.pie(bebe, labels=bebe.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuio de Quem Bebe e No Bebe')
plt.show()
```

### Grfico de Pizza Para Quem Bebe e Fuma

```python
# Criando uma nova coluna para categorizar quem bebe e fuma
df_pesquisa_medica['Bebe e Fuma'] = df_pesquisa_medica.apply(
    lambda row: 'Bebe e Fuma' if row['Bebe'] != 'No' and row['Fuma'] != 'No' 
    else 'S Bebe' if row['Bebe'] != 'No' and row['Fuma'] == 'No'
    else 'S Fuma' if row['Bebe'] == 'No' and row['Fuma'] != 'No'
    else 'No Bebe e No Fuma', axis=1)

# Contagem das categorias
bebe_fuma = df_pesquisa_medica['Bebe e Fuma'].value_counts()

# Grfico de Pizza
plt.figure(figsize=(6, 6))
plt.pie(bebe_fuma, labels=bebe_fuma.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuio de Quem Bebe e Fuma')
plt.show()
```
