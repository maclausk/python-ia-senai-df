```python
# Importando a biblioteca Pandas para análise de dados e atribuindo a ela uma alias 'pd'
import pandas as pd

# importando os dados do arquivo csv '' para um Dataframe 'df_pm2'
df_pm2 = pd.read_csv('dataset/pesquisamusicalv2.csv')

# exibindo os primeiros registros do dataframe
df_pm2.head()
```

```python
# Visualizando Informações Técnicas Sobre o Dataframe
df_pm2.info()
```

```python
# Visualizando Estatísticas Descritivas do Dataframe
df_pm2.describe()
```

```python
# Visualizando as Colunas do DataFrame
df_pm2.columns
```

```python
# Renomeando as Colunas do Dataframe e exibindo-a ao final.
df_pm2.columns = [
    'datahora',
    'rock',
    'hiphop',
    'mpb',
    'reggae',
    'funk',
    'sertanejo',
    'gospel',
    'e-music',
    'samba',
    'forró',
    'trap',
    'professor',
    'faixa_etaria']

df_pm2
```

```python
# Removendo a coluna 'datahora', vinculado o resultado a um novo dataframe 'pesquisa_musical' e exibindo as primeiras linhas do novo dataframe
pesquisa_musical = df_pm2.drop(columns=['datahora'])
pesquisa_musical
```

```python
# Criar um Dataframe apenas para os estilos musicais
df_estilos = df_pm2[['rock','hiphop','mpb','reggae','funk','sertanejo','gospel','e-music','samba','forró','trap']]
```

```python
# Quais DataFrames temos até o momento?

# df_pm2 #DataFrame que contem todas as colunas originais, apenas com a coluna referente a data e hora renomeada para 'datahora'
# pesquisa_musical # Dataframe contendo as colunas originais com exceção da coluna 'datahora' que foi dropada.
# df_estilos # Dataframe que contem apenas as colunas referentes a estilos musicais.
```

```python
# Visualizando as descritivas estatísticas do DataFrame 'df_estilos'
df_estilos.describe()
```

```python
# Somar valores da tabela
df_estilos_soma = df_estilos[['rock','hiphop','mpb','reggae','funk','sertanejo','gospel','e-music','samba','forró','trap']].sum()
df_estilos_soma
```

```python
# Importar a biblioteca matplotlib.pyplot para visualização de dados
import matplotlib.pyplot as plt

# Plotar o conteúdo do DataFrame 'df_estilos_soma' em um gráfico de barras
# - 'kind="bar"' especifica que o gráfico será de barras.
# - 'figsize=(10,6)' define o tamanho da figura (10 de largura por 6 de altura).
# - 'color="black"' define a cor das barras como preta.
df_estilos_soma.plot(kind='bar', figsize=(10, 6), color='black')

# Definir o título do gráfico
plt.title('Quantidade de Ocorrências Por Estilo Musical')

# Definir o rótulo do eixo x
plt.xlabel('Estilos Musicais')

# Definir o rótulo do eixo y
plt.ylabel('Quantidade')

# Rotacionar os rótulos no eixo x em 45 graus para melhor legibilidade
plt.xticks(rotation=45)

# Mostrar o gráfico
plt.show()
```

```python
import pandas as pd
import numpy as np

# Criação de uma Series do Pandas chamada 'df_estilos', que é uma estrutura de dados unidimensional em Pandas, semelhante a uma lista ordenada ou uma coluna em uma tabela.
# Pode ser usada para análises ou visualizações futuras, como criar gráficos para comparação entre os estilos.

df_estilos = pd.Series({
    # Estilo musical: 'rock'
    'rock': 3.265000,

    # Estilo musical: 'hihop'
    'hihop': 3.065000,

    # Estilo musical: 'mpb'
    'mpb': 3.160000,

    # Estilo musical: 'reggae'
    'reggae': 2.545000,

    # Estilo musical: 'funk'
    'funk': 2.410000,

    # Estilo musical: 'sertanejo'
    'sertanejo': 2.810000,

    # Estilo musical: 'gospel'
    'gospel': 2.720000,

    # Estilo musical: 'eletronica'
    'e-music': 3.165000,

    # Estilo musical: 'samba'
    # O valor np.nan (Not a Number) representa que ainda não há valores atribuídos ao estilo 'samba'
    'samba': np.nan,

    # Estilo musical: 'forró'
    # O valor np.nan representa que ainda não há valores atribuídos ao estilo 'forró'
    'forró': np.nan,

    # Estilo musical: 'trap'
    # O valor np.nan representa que ainda não há valores atribuídos ao estilo 'trap'
    'trap': np.nan
})

# A Series foi criada manualmente para armazenar as médias das notas dos estilos musicais.

# Calcule a média dos valores da Series 'df_estilos', ignorando os valores NaN
media = df_estilos.mean()

# Exibe a média calculada, formatada com duas casas decimais
print(f'A média de valores dos estilos musicais é {media:.2f}.')
```

```python
# Preencher os valores ausentes na Series com o valor médio dos estilos musicais '2.89' e ao final mostrar a tabela e valores preenchida. 

df_estilos_preenchidos = df_estilos.fillna(2.89)
df_estilos_preenchidos.head(10)
```

```python
# informações técnicas do Dataframe
df_estilos_preenchidos.info()
```

```python
df_estilos_preenchidos_soma = df_estilos_preenchidos.sum()
```

```python
df_estilos_preenchidos_soma.plot(kind='bar',figsize=(10, 6), color='skyblue')
plt.title('Quantidade de Ocorrências por Estilo Musical')
plt.xlabel('Estilos Musicais')
plt.ylabel('Quantidade')
plt.xticks(rotation=45)
plt.show()
```
