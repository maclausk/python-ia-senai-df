##### Passo 4: Criar Ambiente Virtual

- Um ambiente virtual isola as dependncias do seu projeto, evitando conflitos entre diferentes verses de pacotes. 
- No VSCode, use a barra de pesquisa para criar um ambiente virtual ou use o comando "Control+Shift+P", onde deve ser escolhida a opo '.venv' e em seguida a linguagem Python. 

Para ativar o ambiente virtual: Execute o comando 'activate' no terminal, via prompt de comando, seguindo o seguinte exemplo: '\.venv\Scripts\activate'.



Para desativar o ambiente virtual: Execute o comando 'activate' no terminal, via prompt de comando, seguindo o seguinte exemplo: '\.venv\Scripts\activate'.

##### 7: Observaes Importantes

**Notebooks:** So como cadernos digitais onde voc pode escrever e executar cdigo passo a passo. Eles ajudam a combinar explicaes e grficos em um s lugar, tornando o aprendizado e a anlise de dados mais fceis e interativos.

**Por que usar um ambiente virtual?**
- **Isolamento**: Garante que os pacotes e suas verses sejam especficos para cada projeto, evitando problemas com outros projetos.
- **Organizao**: Facilita a gesto de dependncias, mantendo seu ambiente de desenvolvimento limpo e organizado.

**Por que desativar o ambiente virtual?**
- **Recursos**: Libera memria e processamento quando no est usando o projeto.
- **Flexibilidade**: Permite alternar para outros projetos que podem ter suas prprias configuraes.
- **Consistncia**: Assegura que o projeto use as mesmas verses de pacotes sempre que voc trabalhar nele.
- **Evitar Conflitos**: Previne problemas com pacotes de outras partes do sistema.
