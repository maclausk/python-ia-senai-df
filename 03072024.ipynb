### Aula - 03/07/2024 - Git / Github

#### Git
Git é um sistema de controle de versões que permite salvar e acompanhar mudanças no código.

#### GitHub
GitHub é uma plataforma que hospeda repositórios Git, facilitando o compartilhamento e a colaboração em projetos. Manter os arquivos de aulas diárias no GitHub ajuda a organizar, versionar e apresentar seu progresso, essencial para construir um portfólio e colaborar com outros desenvolvedores.

#### Programa
Um programa é um conjunto de instruções para resolver um problema. Esse programa é composto por algoritmos.

#### Algoritmos
Algoritmos são sequências lógicas de instruções para resolver um problema.

#### Uso de Parênteses
O que está incluso entre parênteses é exatamente o que a função deve executar.

#### Repositórios
Um repositório é um espaço onde os arquivos de um projeto são armazenados e organizados. Ele guarda todo o histórico de alterações, facilitando o gerenciamento e a colaboração em projetos de programação. No GitHub, repositórios permitem que você compartilhe seu trabalho e colabore com outros desenvolvedores.

## Explicação Inicial do Git

Criei o projeto localmente e preciso criar cópias pois o projeto é grande. Se um erro ocorrer e quebrar meu projeto, eu terei backups, conhecidos como versionamento. Cada vez que salvo o projeto, estou criando versões diferentes.

No VSCode, terminal ou PowerShell, para criar o repositório local do Git, eu digito o comando:

```sh
git init
```

Pronto, repositório local criado. O nome "master" significa que essa é a minha primeira versão do projeto.

Agora eu vou desenvolvendo meu projeto, e a cada etapa concluída, salvo as mudanças.

O tronco da nossa árvore de projeto é o principal, e apenas o scrum master tem acesso ao projeto principal. Eu recebo um galho, uma branch, que é uma cópia do projeto para eu trabalhar. Não mexo diretamente no original. Após fazer alterações, envio minha cópia de volta para o projeto principal com minhas mudanças.

Para salvar uma versão do projeto, usamos:

```sh
git commit -m "primeiro commit"
```

O `-m` significa que vamos colocar uma mensagem, como por exemplo, "ajuste na tela de login". Assim, foi criada uma nova versão dentro do galho principal, o master.

Para criar uma nova branch, usamos:

```sh
git branch Marcelo
```

Isso cria uma nova branch chamada "Marcelo". Outro usuário deve usar essa branch para fazer alterações, usando:

```sh
`git checkout Marcelo`
```

Os projetos serão salvos na branch "Marcelo". O comando `git checkout` diz que quero mudar de galho.

```sh
git checkout
```

Quando todos terminarem suas partes, podemos juntar todos os galhos no projeto principal por meio do comando:

```sh
`git merge`
```

Isso junta o que os usuários fizeram com o que já estava no galho principal. 

Ao juntar os projetos, podem ocorrer conflitos, que precisam ser validados linha por linha. O comando:

```sh
`git merge usuario`
```

significa que, estando na branch principal, estou pedindo o merge das alterações do usuário para a branch principal. Após validar tudo, juntamos tudo na branch principal e enviamos os arquivos para a nuvem com o comando:

```sh
`git push origin master`
```

### Sincronizando o diretório local com o Github ###

- Agora eu quero encaminhar os arquivos que estavam localmente para a nuvem, deixando ambos os ambientes idênticos.
- No diretório local onde você fez as alterações ou criou novos arquivos, use o Git Bash para adicionar os arquivos ao repositório local:

```sh
git add .
```

- Este comando adiciona todos os arquivos modificados e novos ao índice (staging area) do Git, preparando-os para o commit
- Em seguida, faça um commit das alterações com uma mensagem descritiva:

```sh
git commit -m "Adicionando novos arquivos e atualizando o projeto"
```
- O comando git commit cria um snapshot das alterações preparadas. A opção -m permite adicionar uma mensagem descritiva, que explica as mudanças feitas.
- Finalmente, envie as alterações para o repositório no GitHub:

```sh
git push origin main
```
- O comando git push envia seus commits para o repositório remoto no GitHub. origin é o nome padrão para o repositório remoto, e main é o branch principal.
- Após esses passos, os arquivos do seu diretório local serão enviados para o repositório no GitHub, mantendo ambos os ambientes sincronizados.

````python
# Demais Operadores Matemáticos

# Solicitando a entrada de dois números pelo usuário
n1 = input("Digite o primeiro número: ")
n2 = input("Digite o segundo número: ")

# Convertendo os números para inteiro
# Na entrada de dados acima, os números são inseridos como strings pelo comando input.
# Isso faz parte das melhores práticas para se codificar, garantindo que estamos trabalhando com o tipo de dado correto.

n1 = int(n1)
n2 = int(n2)

# Realizando operações matemáticas
divisao_inteira = n1 // n2    # Divisão inteira
divisao = n1 / n2             # Divisão normal
resto = n1 % n2               # Resto da divisão (verificação de paridade)
expo = n1 ** n2               # Exponenciação

# Exibindo os resultados
print(f'Divisão Inteira: {divisao_inteira}')
print(f'Divisão normal: {divisao}')
print(f'Resto: {resto} (se o resto for 0, o número é par; se for diferente de 0, o número é ímpar)')
print(f'Exponenciação: {expo}')

# Explicação do código:
# 1. Solicitamos dois números ao usuário usando o comando `input`.
# 2. Convertendo essas entradas de string para inteiros usando `int()`.
# 3. Realizamos várias operações matemáticas:
#    - Divisão inteira (`//`): Divide dois números e arredonda para o inteiro mais próximo.
#    - Divisão normal (`/`): Divide dois números normalmente, resultando em um número flutuante.
#    - Resto da divisão (`%`): Obtém o resto da divisão dos dois números, útil para verificar se um número é par ou ímpar.
#    - Exponenciação (`**`): Eleva um número à potência de outro.
# 4. Usamos `f-strings` para exibir os resultados de forma clara e formatada.
````

````python
# Atribuição de Subtração

valor = 10  # Atribuímos o valor 10 à variável 'valor'

# Subtraímos 10 do valor atual de 'valor' usando o operador de atribuição -=
valor -= 10  # Isso é equivalente a escrever: valor = valor - 10

# Exibindo o resultado
print(f'Resultado de atribuição de subtração: {valor}')

# Explicação do código:
# 1. Inicialmente, a variável 'valor' é atribuída com o valor 10.
# 2. Em seguida, utilizamos o operador de atribuição de subtração ( -= ) para subtrair 10 do valor atual da variável 'valor'.
#    - Este operador atualiza a própria variável, subtraindo o valor especificado (neste caso, 10).
#    - É o mesmo que escrever: valor = valor - 10
# 3. Por fim, o valor atualizado de 'valor' é exibido usando uma f-string.
````

````python
# Atribuição de Soma

valor = 0  # Inicializamos a variável 'valor' com 0

# Utilizamos o operador de atribuição de soma ( += ) para adicionar 10 ao valor atual da variável 'valor'
valor += 10  # Isso é equivalente a escrever: valor = valor + 10

# Exibindo o resultado
print(f'Resultado de atribuição de soma: {valor}')

# Explicação do código:
# 1. Inicializamos a variável 'valor' com 0.
# 2. Em seguida, utilizamos o operador de atribuição de soma ( += ) para adicionar 10 ao valor atual da variável 'valor'.
#    - Este operador atualiza a própria variável, somando o valor especificado (neste caso, 10).
#    - É o mesmo que escrever: valor = valor + 10
# 3. Por fim, o valor atualizado de 'valor' é exibido usando uma f-string.

# Nota: A frase "o código vai se repetir até que a exceção seja executada" não se aplica aqui,
# pois este código executa uma única vez. Para repetição, seria necessário utilizar um loop.
````

````python
# Atribuição de Multiplicação

valor = 1  # Inicializamos a variável 'valor' com 1

# Utilizamos o operador de atribuição de multiplicação ( *= ) para multiplicar o valor atual da variável 'valor' por 10
valor *= 10  # Isso é equivalente a escrever: valor = valor * 10

# Exibindo o resultado
print(f'Resultado de atribuição de multiplicação: {valor}')

# Explicação do código:
# 1. Inicializamos a variável 'valor' com 1.
# 2. Em seguida, utilizamos o operador de atribuição de multiplicação ( *= ) para multiplicar o valor atual da variável 'valor' por 10.
#    - Este operador atualiza a própria variável, multiplicando pelo valor especificado (neste caso, 10).
#    - É o mesmo que escrever: valor = valor * 10
# 3. Por fim, o valor atualizado de 'valor' é exibido usando uma f-string.
````

````python
# Atribuição de Divisão

valor = 100  # Inicializamos a variável 'valor' com 100

# Utilizamos o operador de atribuição de divisão ( /= ) para dividir o valor atual da variável 'valor' por 10
valor /= 10  # Isso é equivalente a escrever: valor = valor / 10

# Exibindo o resultado
print(f'Resultado de atribuição de divisão: {valor}')

# Explicação do código:
# 1. Inicializamos a variável 'valor' com 100.
# 2. Em seguida, utilizamos o operador de atribuição de divisão ( /= ) para dividir o valor atual da variável 'valor' por 10.
#    - Este operador atualiza a própria variável, dividindo pelo valor especificado (neste caso, 10).
#    - É o mesmo que escrever: valor = valor / 10
# 3. Por fim, o valor atualizado de 'valor' é exibido usando uma f-string.
````

````python
# Atribuição de Divisão Inteira

valor = 100  # Inicializamos a variável 'valor' com 100

# Utilizamos o operador de atribuição de divisão inteira ( //= ) para dividir o valor atual da variável 'valor' por 10
# e obter o resultado inteiro da divisão
valor //= 10  # Isso é equivalente a escrever: valor = valor // 10

# Exibindo o resultado
print(f'Resultado de atribuição de divisão inteira: {valor}')

# Explicação do código:
# 1. Inicializamos a variável 'valor' com 100.
# 2. Em seguida, utilizamos o operador de atribuição de divisão inteira ( //= ) para dividir o valor atual da variável 'valor' por 10
#    e obter o resultado inteiro da divisão.
#    - Este operador atualiza a própria variável, realizando a divisão inteira pelo valor especificado (neste caso, 10).
#    - É o mesmo que escrever: valor = valor // 10
# 3. Por fim, o valor atualizado de 'valor' é exibido usando uma f-string.
````
