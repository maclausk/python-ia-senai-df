### Entendendo Branches no Git

No Git, as branches (ou ramificações) são uma forma de gerenciar diferentes linhas de desenvolvimento no seu projeto. Isso permite que você trabalhe em novas funcionalidades ou correções de bugs sem afetar o código principal até que esteja pronto para ser integrado.

#### O que é a Branch Principal?
- **Main**: É a branch principal do seu repositório. Toda vez que você inicia um projeto no Git, a branch principal é criada automaticamente e, por padrão, é chamada de `main`. É a versão "oficial" do seu projeto, onde geralmente o código mais estável e finalizado é mantido.

#### O que são Branches Secundárias?
- **Branches Secundárias**: São outras ramificações que você pode criar a partir da branch principal (ou de qualquer outra branch) para desenvolver novas funcionalidades, testar ideias ou corrigir problemas. As branches secundárias são independentes da `main` até que você decida mesclá-las de volta.

#### Como Criar e Trabalhar com Branches?

1. **Criar uma Branch**:
- Quando você precisa desenvolver uma nova funcionalidade ou corrigir um bug, é uma boa prática criar uma nova branch a partir da `main`. Isso mantém o código principal intacto enquanto você faz as alterações necessárias.

```bash
git checkout -b minha-nova-branch
```

Nesse comando:

checkout muda para a nova branch.
-b cria uma nova branch com o nome que você escolheu (minha-nova-branch).

2. **Fazendo Modificações:**
- Com a nova branch criada, você pode fazer as modificações que precisa. As alterações que você fizer só irão afetar essa nova branch, e não a main.

```bash
git add .
git commit -m "Descrição das mudanças"
```

- Aqui, `git add .` prepara as mudanças, e git commit as salva na branch atual.

3. **Voltando para a Branch Principal:**
- Depois de terminar as modificações e testar que tudo está funcionando, você pode voltar para a branch main:

```bash
 git checkout main
```
- Agora, se as alterações estiverem prontas para serem integradas ao código principal, você pode mesclar a branch secundária com a main:

```bash
git merge minha-nova-branch
```
- Isso integra as mudanças feitas na minha-nova-branch à main, atualizando o código principal com suas novas funcionalidades ou correções.

**Exemplo Prático**
- Imagine que você está desenvolvendo um site e precisa adicionar uma nova funcionalidade de login. Em vez de fazer isso diretamente na main, você cria uma nova branch chamada feature-login para desenvolver e testar essa funcionalidade.
- Depois que a funcionalidade estiver funcionando corretamente, você pode voltar para a main e mesclar as mudanças, garantindo que o código principal só receba código que foi completamente testado.

**Vantagens de Usar Branches**
**Segurança:** Mantém o código principal seguro e estável enquanto você trabalha em novas funcionalidades.
**Organização:** Permite organizar o trabalho de forma mais clara, especialmente em equipes onde várias funcionalidades estão sendo desenvolvidas simultaneamente.
**Facilidade de Reversão:** Se algo der errado, é fácil descartar as mudanças na branch secundária sem afetar o código principal.

<i>Branches são uma parte essencial do trabalho com Git e ajudam a garantir que o desenvolvimento do seu projeto seja organizado, seguro e eficiente.</i>

### Como Criar um Ambiente Virtual (.venv) no Python

Para isolar as dependências do seu projeto Python e evitar conflitos com outras bibliotecas instaladas globalmente, é uma boa prática criar um ambiente virtual. Um ambiente virtual é uma espécie de "contenedor" onde você pode instalar pacotes específicos para o seu projeto, sem afetar o ambiente global do Python.

#### Criando um Ambiente Virtual (.venv)

1. **Abra o Terminal ou Prompt de Comando**:
   - Navegue até a pasta do seu projeto.

2. **Execute o Comando para Criar o Ambiente Virtual**:
   - No terminal, execute o seguinte comando:

```bash
python -m venv venv
```

**Explicação:**
- `python -m venv`: Utiliza o módulo venv do Python para criar um ambiente virtual.
- **venv**: Esse é o nome do diretório onde o ambiente virtual será criado. Você pode nomeá-lo como quiser, mas venv é uma convenção comum.

3. **Ativando o Ambiente Virtual:**
- Após a criação, você precisa ativar o ambiente virtual para começar a usá-lo.

```bash
.\venv\Scripts\activate
```

4. **Instalando Pacotes no Ambiente Virtual:**
- Com o ambiente virtual ativado, qualquer pacote que você instalar usando pip será instalado apenas nesse ambiente, e não globalmente.

```bash
pip install nome-do-pacote
```

5. **Desativando o Ambiente Virtual:**

- Quando terminar de trabalhar no seu projeto, você pode desativar o ambiente virtual com o comando:

```bash
deactivate
```

6. **Benefícios do Uso de um Ambiente Virtual**
- **Isolamento:** Assegura que as dependências de um projeto não afetem outros projetos.
- **Controle de Dependências:** Facilita o gerenciamento das bibliotecas e versões específicas que seu projeto requer.
- **Portabilidade:** Ao compartilhar seu projeto, outros desenvolvedores podem recriar o mesmo ambiente usando um arquivo de dependências, como requirements.txt.

<i>Criar e usar ambientes virtuais é uma prática recomendada em todos os projetos Python, garantindo maior controle e organização no desenvolvimento.</i>

#### Tipos de Conjuntos em Python

- `[]` - **List**: Uma lista é uma coleção ordenada de elementos que podem ser alterados. Exemplo: `[1, 2, 3]`
- `{}` - **Set**: Um conjunto é uma coleção desordenada de elementos únicos. Exemplo: `{1, 2, 3}`
- `{chave:valor}` - **Dict**: Um dicionário é uma coleção não ordenada de pares de chave e valor. Exemplo: `{'nome': 'João', 'idade': 25}`
- `()` - **Tuple**: Uma tupla é uma coleção ordenada de elementos que não podem ser alterados. Exemplo: `(1, 2, 3)`

````python
# Definir a coluna 'País' como o índice do DataFrame 'df_ic'
# Isso facilita o acesso e filtragem dos dados por país, como por exemplo, filtrar pelo país 'Brasil'.
# O argumento 'inplace=True' faz a alteração diretamente no DataFrame original, alterando sua estrutura.
df_ic.set_index('País', inplace=True)

# Mostrar os primeiros registros do DataFrame para verificar as alterações
df_ic.head()
````

````python
# Utilizando a função map para transformar um intervalo de números em strings


# 'Iteração', 'Looping', e 'Repetição' referem-se ao processo de passar por uma sequência de elementos (como números, strings, etc.) e executar uma operação em cada um deles.
# Criar a variável chamada 'anos' e utilizar a função map para aplicar a função 'str' a cada item no intervalo de 1980 a 2013, convertendo-os para strings.
# Em seguida, o resultado é convertido em uma lista com a função 'list', com o nome de 'anos', armazenando os anos como strings.
anos = list(map(str, range(1980, 2014)))

# Exibir a lista de anos
anos
````

#### Extrair a Série de dados da coluna 'Brasil' usando loc onde 'anos' é uma lista que contém os anos de 1980 a 2013.

- Exibir a Série extraída
- Transformar a Série em um dicionário
- Exibir o dicionário
- Transformar o dicionário em um DataFrame
- Exibir o DataFrame resultante

#### Explicação do Código:
- loc[] vs iloc[]:
    - loc[]: Utilizado para localizar dados pelo label (rótulo ou nome) das linhas ou colunas.
    - iloc[]: Utilizado para localizar dados pelo índice (posição numérica) das linhas ou colunas.

````python
# Extrair a serie de dados do índice 'Brasil' e da lista 'anos'
brasil = df_ic.loc['Brasil',anos]
brasil
````

````python
# Verificando o tipo da variável 'brasil'
type(brasil)
````

````python
# Criar um Dicionário através da Série 'brasil'
# Dicionário possui 'chave:valor'. Utilizando 'brasil.index' você descobre o índice e 'brasil.values' você descobre o valor.
brasil_dict = {'ano': brasil.index.tolist(),  # Lista de anos a partir dos índices da Série
               'imigrantes': brasil.values.tolist()}  # Lista de valores de imigrantes a partir dos valores da Série

# Apresentando o dicionário
brasil_dict
````

````python
# Transformar o dicionário 'brasil_dict' em um dataframe somente com valores do 'brasil'
df_brasil = pd.DataFrame(brasil_dict)

# Exibir o DataFrame
df_brasil
````

````python
# Criando o gráfico de linhas
# O método 'plt.plot()' é usado para criar gráficos de linhas.
# Os dados do eixo x (anos) são extraídos da coluna 'ano' do DataFrame 'df_brasil'.
# Os dados do eixo y (quantidade de imigrantes) são extraídos da coluna 'imigrantes' do DataFrame 'df_brasil'.
plt.plot(df_brasil['ano'], df_brasil['imigrantes'])

# Rotacionando os rótulos do eixo x em 45 graus para melhorar a legibilidade
plt.xticks(rotation=45)

# Dando título ao gráfico
plt.title('Imigrantes Brasileiros no Canadá (1980-2013)')  

# Definindo rótulos específicos para o eixo x
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015'])
plt.yticks([500,1000,1500,2000,2500,3000]) # valores

# Incluindo grades no gráfico
plt.grid()

# Salvando o gráfico como uma imagem PNG
# Vantagem: Aumentar o potencial dos comandos.
plt.savefig('Material de Apoio\grafico_imigrantes_brasil.png')
plt.figure(figsize=(8,4))
# Qual é a diferença entre os dois comandos acima?

# 1 polegada = 2.54 cm
# 8 polegadas de largura x 4 polegadas de altura.
    # largura 8x2.54cm = 20.32 cm
    # altura  4x2.54cm = 10.16 cm
# DPI (dots per inch, ou pontos por polegada).
    # Padrão no matplotib - DPI = 100
    # Largura : 8 polegadas x 100 DPI = 800 pixels.
    # Altura  : 4 polegadas x 100 DPI = 400 pixels.

# Mostrando o gráfico
plt.show()
````

````python
# Criar a figura e os eixos para os subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))  
# Cria uma figura com dois subplots lado a lado (1 linha e 2 colunas).
# `fig` é a figura que contém os gráficos.
# `axes` são os subplots dentro da figura onde os gráficos serão desenhados.
# `figsize=(14, 6)` define o tamanho da figura (14 de largura por 6 de altura).

# Gráfico de linhas no subplot à esquerda (axes[0])
axes[0].plot(df_brasil['ano'], df_brasil['imigrantes'], color='blue')  
# Cria um gráfico de linhas no primeiro subplot (à esquerda).
# `df_brasil['ano']` são os valores no eixo x (anos).
# `df_brasil['imigrantes']` são os valores no eixo y (número de imigrantes).
# `color='blue'` define a cor da linha do gráfico.

axes[0].set_title('Imigrantes Brasileiros no Canadá (1980-2013)')  
# Define o título do gráfico no primeiro subplot (à esquerda).

axes[0].set_xlabel('Ano')  
# Define o rótulo do eixo x no gráfico à esquerda.

axes[0].set_ylabel('Número de Imigrantes')  
# Define o rótulo do eixo y no gráfico à esquerda.

axes[0].set_xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010', '2015'])  
# Define rótulos específicos para o eixo x, mostrando os anos desejados.

axes[0].grid(True)  
# Ativa a grade no gráfico à esquerda, facilitando a visualização dos dados.

# Boxplot no subplot à direita (axes[1])
axes[1].boxplot(df_brasil['imigrantes'])  
# Cria um boxplot no segundo subplot (à direita) para mostrar a distribuição dos imigrantes brasileiros.
# `df_brasil['imigrantes']` são os valores que serão representados no boxplot.

axes[1].set_title('Distribuição do Número de Imigrantes')  
# Define o título do gráfico no segundo subplot (à direita).

axes[1].set_xlabel('Brasil')  
# Define o rótulo do eixo x no gráfico à direita.

axes[1].set_ylabel('Número de Imigrantes')  
# Define o rótulo do eixo y no gráfico à direita.

# Ajustar o layout para não sobrepor os gráficos
plt.tight_layout()  
# Ajusta o layout da figura para garantir que os subplots não se sobreponham ou fiquem muito próximos.

# Salvando a figura como uma imagem PNG
plt.savefig('Material de Apoio/grafico_imigrantes_brasil_e_boxplot.png')  
# Salva a figura contendo os dois gráficos em um arquivo de imagem PNG.
# O caminho e o nome do arquivo são especificados como 'Material de Apoio/grafico_imigrantes_brasil_e_boxplot.png'.

# Mostrar a figura com os dois gráficos
plt.show()  
# Exibe a figura na tela com os dois gráficos (o gráfico de linhas e o boxplot).
````

### Storytelling sobre os Gráficos de Imigrantes Brasileiros no Canadá

Imagine que estamos explorando a jornada de brasileiros que decidiram deixar o Brasil em busca de novas oportunidades no Canadá ao longo das últimas décadas. Essa jornada, capturada em números, revela não apenas a quantidade de pessoas que fizeram essa escolha, mas também as tendências e mudanças ao longo do tempo.

#### Gráfico de Linhas: Evolução dos Imigrantes Brasileiros no Canadá (1980-2013)

O primeiro gráfico, uma linha que serpenteia ao longo dos anos de 1980 a 2013, conta a história dessa migração. Nos primeiros anos, vemos um movimento modesto, com menos de 500 brasileiros se mudando para o Canadá anualmente. Era um período de descoberta, onde apenas alguns pioneiros cruzavam as fronteiras em busca de uma vida melhor.

No entanto, à medida que nos aproximamos dos anos 1990 e 2000, a linha começa a subir. O crescimento é visível, indicando que mais brasileiros estavam sendo atraídos pela promessa de estabilidade e oportunidades no Canadá. Esse aumento pode estar relacionado a diversos fatores, como crises econômicas no Brasil, políticas de imigração mais abertas no Canadá, ou simplesmente a expansão das redes de apoio entre imigrantes.

O pico no gráfico, que ocorre por volta de 2010, sugere um ápice na imigração brasileira. Durante esse período, o número de imigrantes ultrapassa os 2500 por ano, um sinal de que o Canadá havia se tornado um destino atraente e acessível para muitos brasileiros. No entanto, após esse pico, vemos uma queda acentuada, sugerindo uma mudança nas condições ou nas políticas que influenciavam a imigração.

O segundo gráfico, um boxplot, oferece uma visão diferente e mais técnica dessa história. Ele nos mostra como esses números variaram ao longo dos anos, destacando não apenas os valores medianos, mas também a dispersão e os outliers—os anos em que os números de imigração foram excepcionalmente altos ou baixos.

No boxplot, a "caixa" central representa a metade dos dados, concentrando-se em torno da mediana. A posição dessa caixa e o comprimento de seus "bigodes" indicam a variabilidade dos dados. No caso da imigração brasileira, podemos ver que a maioria dos anos se concentrou em um intervalo relativamente estreito, mas com alguns anos em que os números de imigração foram significativamente maiores, refletindo os picos observados no gráfico de linhas.

Este gráfico também destaca a estabilidade relativa na imigração ao longo do tempo, com algumas exceções marcadas. Esses outliers podem ser interpretados como anos extraordinários, onde eventos específicos ou mudanças nas políticas influenciaram significativamente o número de brasileiros imigrando para o Canadá.
